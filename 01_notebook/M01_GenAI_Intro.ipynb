# 1 | Überblick GenAI
---

Generative Künstliche Intelligenz (GenAI) bezeichnet KI-Modelle, die in der Lage sind, **eigenständig** **neue Inhalte** zu erzeugen – dazu zählen Texte, Bilder, Audio- und Videoformate. Diese Modelle basieren meist auf sogenannten **Foundation Models**, insbesondere **Large Language Models (LLMs)** wie GPT oder Claude, die durch das Training auf großen Datenmengen gelernt haben, Muster zu erkennen und kreative Inhalte zu generieren.



Multimodale Inhalte – Ein Überblick:

* **Text**: Erstellung von Texten, Zusammenfassungen, Übersetzungen, Code und interaktive Dialoge (z. B. Chatbots, Assistenten).
* **Bild**: Generierung von Bildern aus Textbeschreibungen (z. B. DALL·E, Midjourney), Bearbeitung bestehender Bilder (Inpainting/Outpainting), Klassifikation und Objekt-Erkennung.
* **Audio**: Umwandlung von Sprache in Text (Speech-to-Text), umgekehrt (Text-to-Speech), Klanganalysen, Podcasts und Musikkomposition.
* **Video**: Automatische Videoerstellung aus Text, Videoanalyse, Bild-zu-Video-Transformation und Erkennung von Inhalten in Videos.

Im Laufe dieses Kurses werden sowohl textbasierte als auch multimodale Modelle konzipiert und kritisch reflektiert. Ziel ist es, ein fundiertes Verständnis für die Potenziale und Grenzen generativer KI-Technologien zu entwickeln.

Problemklassen
</font></p>

Diese Tabelle dient als Orientierungshilfe. Sie beantwortet nicht die Frage „Welches Tool ist das beste?“, sondern die wichtigere Frage:

> *Welche Art von Problem habe ich – und welcher Lösungsweg passt grundsätzlich dazu?*

In der Praxis scheitern viele KI-Projekte nicht am Modell, sondern daran, dass Chat, Workflow, Code oder Agenten für das falsche Problem eingesetzt werden. Die Tabelle ordnet typische GenAI-Szenarien nach Problemklassen und zeigt einen sinnvollen Startpunkt, nicht die endgültige Architektur.

| Problemklasse | Typisches Szenario | Woran erkennt man das? | Empfohlener Start | Typischer Anfängerfehler |
|---|---|---|---|---|
| Kreativer Assistent | „Ich brauche Ideen, Texte oder Erklärungen.“ | Einmalig, interaktiv, Mensch denkt mit | Chat / Copilot | Zu früh automatisieren oder integrieren wollen |
| Prozess-Automatisierer | „Das Gleiche passiert immer wieder nach einem klaren Auslöser.“ | Wiederholbar, fester Trigger, wenig Logik | Workflow (Make / n8n) | Workflows für komplexe Logik missbrauchen |
| Daten-Verarbeiter | „Viele Dateien oder Datensätze müssen systematisch ausgewertet werden.“ | Große Menge, Schleifen, Kosten relevant | Python + API | Alles per Chat oder Workflow lösen wollen |
| Autonomer Entwickler | „Die KI soll selbstständig mit Code oder Dateien arbeiten.“ | Kontext über mehrere Schritte, Exploration | Agent, Python + API | Agenten einsetzen, wo ein Script reicht |
| App-Entwickler (intern) | „Andere sollen das regelmäßig nutzen.“ | UI nötig, Wissensbasis, stabiler Betrieb | KI-App-Builder | Chat als Produkt missverstehen |
| Datenschützer | „Ich muss Patientendaten analysieren, die das Haus nicht verlassen dürfen.“ | Sensible Daten, rechtliche Vorgaben, kein Cloud-Zugriff | Lokale LLMs (z. B. Ollama), <br>Python + API | Datenschutz nur als Hosting-Frage sehen |

<p><font color='black' size="5">
OpenAI Konversationsformat
</font></p>

OpenAI bietet ein Konversationsformat für die Nutzung seiner Sprachmodelle, insbesondere für die Interaktion mit Chatbot-artigen Anwendungen. Dieses Format wird typischerweise in der ChatCompletion-API verwendet, die speziell für dialogorientierte Aufgaben entwickelt wurde. Die API erleichtert die Strukturierung von Dialogen, indem sie die Verwendung von rollenbasierten Nachrichten unterstützt, um den Kontext und die Intentionen der Beteiligten klar zu definieren.

Grundprinzipien des Konversationsformats von OpenAI:   
In der Chat-API von OpenAI besteht ein Dialog aus einer Serie von Nachrichten, die jeweils einer Rolle zugeordnet sind. Die Hauptrollen sind:

+ **system**: Definiert den **Kontext** oder die Anweisungen für das Modell.
+ **user**: Enthält die **Benutzer**eingaben.
+ **assistant**: Die vom **Modell** generierten Antworten.

```python
from openai import OpenAI, chat

client = OpenAI()

response = chat.completions.create(
    model="gpt-5.4-nano",
    messages=[
        {"role": "system", "content": "Du bist ein hilfreicher KI-Assistent."},
        {"role": "user", "content": "Generative KI ist ..."},
    ]
)

mprint('## 🤖 KI:')
mprint(response.choices[0].message.content)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Laden eines Modells für Übersetzung (Englisch → Deutsch)
model_name = "Helsinki-NLP/opus-mt-en-de"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Text zum Übersetzen
text_to_translate = "Why is the sky blue?"

# Übersetzung durchführen
inputs = tokenizer(text_to_translate, return_tensors="pt")
outputs = model.generate(**inputs)
translation = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Ergebnis anzeigen
mprint('## 🤖 KI:')
mprint(translation)
```

<img src="https://images.seeklogo.com/logo-png/61/2/langchain-icon-logo-png_seeklogo-611655.png" class="logo" width="180"/>

LangChain ist ein KI-Framework, das die *Entwicklung* von Sprachanwendungen  **vereinfacht**, indem es Werkzeuge zur Integration von Sprachmodellen mit externen Datenquellen und Diensten bietet. Es ist speziell darauf ausgerichtet, die Fähigkeiten von Large Language Models (LLMs) wie OpenAI's GPT-x zu nutzen und diese Modelle effektiv mit anderen Systemen zu verknüpfen, um praktische und funktionale Anwendungen zu erstellen.

Das Framework bietet eine Reihe von Funktionen, darunter:
- **Chain-of-Thought Prompting**: Dies hilft, die Antworten von Sprachmodellen zu verbessern, indem es den Denkprozess strukturiert, der in den Antworten enthalten ist.
- **Retrieval-Augmented Generation (RAG)**: Eine Methode, die es ermöglicht, externe Informationen während der Generierung von Antworten dynamisch abzurufen, um die Genauigkeit und Relevanz der Antworten zu verbessern.
- **Modulare Architektur**: LangChain ermöglicht es Entwicklern, verschiedene Komponenten wie Suchmaschinen, Datenbanken und andere Dienste einfach zu integrieren und zu kombinieren, um komplexe Anwendungen zu erstellen.

Das Framework unterstützt die Erstellung von Anwendungen, die über einfache Textgenerierung hinausgehen und tieferes Verständnis und Interaktionen ermöglichen, indem es die Vorteile von Large Language Models mit spezifischen, aufgabenorientierten Informationen verbindet.


**Das LangChain-Ökosystem** verbindet Modell-Orchestrierung, Agenten-Workflows, Integrationen und Observability in einer Toolchain.

```python
#@markdown   <p><font size="4" color='green'>  LangChain-Ökosystem</font> </br></p>

diagram = """
%%{init: {'theme':'forest'}}%%
graph TD
    %% Definition der Stile für die einzelnen Boxen
    classDef ecosystem fill:#1565c0,stroke:#0d47a1,stroke-width:2px,color:#ffffff;
    classDef core fill:#0d47a1,stroke:#002171,stroke-width:2px,color:#ffffff;
    classDef greyBox fill:#A9A9A9,stroke:#707070,stroke-width:2px,color:#000000;

    subgraph Development ["<b>Entwicklung & Logik</b>"]
        LC_Core["<b>LangChain</b><br/>Basis & LCEL"]
        LC_Comm["<b>langchain-community</b><br/>Integrationen"]
        LC_Partner["<b>Provider-Pakete</b><br/>(OpenAI, etc.)"]
        LG["<b>LangGraph</b><br/>Workflows & Agenten"]
    end

    subgraph Observability ["<b>Qualität & Monitoring</b>"]
        LS["<b>LangSmith</b><br/>Tracing & Testing"]
    end

    subgraph Deployment ["<b>Bereitstellung</b>"]
        LGP["<b>LangSmith Deployment</b><br/>HF-Spaces (kostenlos)"]
    end

    %% Styling der Subgraphs (Hintergrund-Boxen)
    style Development fill:#e3f2fd,stroke:#1565c0,stroke-width:1px,stroke-dasharray: 5 5
    style Observability fill:#f3e5f5,stroke:#7b1fa2,stroke-width:1px,stroke-dasharray: 5 5
    style Deployment fill:#f5f5f5,stroke:#9e9e9e,stroke-width:1px

    %% Verbindungen
    LC_Core --> LC_Comm
    LC_Core --> LC_Partner
    LC_Core --> LG

    LG --> LS
    LC_Core --> LS

    LC_Core --> LGP
    LG --> LGP

    %% Klassen zuweisen
    class LC_Core core
    class LC_Comm,LC_Partner,LG,LS ecosystem
    class LGP greyBox
"""

mermaid(diagram, width=750)
```

<p><font color='black' size="5">
Eine einzelne Frage stellen

Um dem Modell eine einzelne, von vorherigen Dialogen unabhängige Frage zu stellen, kann eine Zeichenfolge als Eingabe verwendet werden, auf die das Modell direkt reagiert.

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("openai:gpt-5.4-nano")

# Definieren der Frage
user_input = "Was sind die fünf grössten Städte in Deutschland?"

# Aufruf des LLMs
response = llm.invoke(user_input)

# Ausgabe der Antwort
mprint("## ✨ Model response:")
mprint("---")
mprint(response.content)
```

<p><font color='black' size="5">
Überblick LangChain Konzepte
</font></p>

Konzepte von LangChain sind grundlegende Bausteine und Prinzipien, die das Framework ausmachen und seine Funktionsweise ermöglichen. Hier sind einige der wichtigsten Konzepte:

+ Chat-Modelle (**Chat models**): LLMs, die über eine Chat-API verfügbar sind und Sequenzen von Nachrichten verarbeiten1.
+ Nachrichten (**Messages**): Kommunikationseinheiten in Chat-Modellen für Ein- und Ausgabe1.
+ Chat-Verlauf (**Chat history**): Eine Sequenz von Nachrichten, die eine Konversation darstellt.
+ Tools (**Tools**): Funktionen mit definierten Schemata für Name, Beschreibung und Argumente.
+ Strukturierte Ausgabe (**Structured output**): Technik, um Chat-Modelle in strukturierten Formaten antworten zu lassen.
+ Retrieval Augmented Generation (**RAG**): Technik zur Verbesserung von Sprachmodellen durch Kombination mit externen Wissensbasen.
+ Prompt-Vorlagen (**Prompt template**s): Komponenten zur Erstellung strukturierter Prompts für LLMs.
+ Chains (**Chains**): Verknüpfungen mehrerer LLMs oder anderer Komponenten für komplexere Anwendungen.
+ Agenten (**Agents**): Nutzen Sprachmodelle, um Aktionssequenzen auszuwählen und mit externen Ressourcen zu interagieren.
+ Retriever (**Retriever**): Komponenten, die relevante Dokumente aus einer Wissensbasis abrufen.
