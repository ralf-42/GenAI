# 1 | Fine-Tuning von Large Language Models (LLMs)
---




**Fine-Tuning** bezeichnet die gezielte Anpassung eines bereits vortrainierten Large Language Models (LLM) an eine bestimmte Aufgabe, Domäne oder einen speziellen Sprachstil. Dabei wird das Modell mit zusätzlichen, auf den Anwendungsfall zugeschnittenen Trainingsdaten weitertrainiert.

Das Vortraining eines LLMs erfolgt auf sehr großen, allgemeinen Textsammlungen. Dadurch verfügt das Modell über umfangreiches Sprach- und Weltwissen. Für spezielle Anforderungen – beispielsweise die Beantwortung von Fachfragen, die Erstellung von Texten in einem bestimmten Stil oder die Unterstützung interner Unternehmensprozesse – kann dieses allgemeine Wissen jedoch nicht immer ausreichen. Durch Fine-Tuning lernt das Modell, die gewünschten Muster, Fachbegriffe und Antwortformate zuverlässiger anzuwenden.

Typische Anwendungsgebiete sind:

* Fachspezifische Chatbots (z. B. Medizin, Recht oder Technik)
* Anpassung an Unternehmenssprache und Corporate Wording
* Klassifikations- und Extraktionsaufgaben
* Generierung strukturierter Antworten
* Verbesserung der Leistung für eng definierte Aufgaben

Fine-Tuning unterscheidet sich von Verfahren wie **Prompt Engineering** oder **Retrieval-Augmented Generation (RAG)**. Während Prompt Engineering die Eingaben optimiert und RAG externe Wissensquellen zur Laufzeit einbindet, werden beim Fine-Tuning die Modellgewichte selbst angepasst. Dadurch kann das Modell bestimmte Aufgaben effizienter und konsistenter ausführen.

Der größte Vorteil des Fine-Tunings liegt in der Spezialisierung des Modells. Allerdings erfordert das Verfahren geeignete Trainingsdaten, Rechenressourcen und eine sorgfältige Qualitätskontrolle, um unerwünschte Effekte wie Überanpassung (Overfitting) zu vermeiden.


# 2 | Parameter Efficient Fine-Tuning
---

Parameter Efficient Fine-Tuning (PEFT) bezeichnet eine Reihe von Techniken, die darauf abzielen, große Sprachmodelle (LLMs) mit deutlich weniger Rechenaufwand und Speicherbedarf anzupassen, als dies beim traditionellen "Full Fine-Tuning" der Fall wäre. Anstatt alle Milliarden von Parametern eines vortrainierten Modells zu aktualisieren, frieren PEFT-Methoden einen Großteil der ursprünglichen Modellgewichte ein und trainieren nur eine kleine Anzahl von neu hinzugefügten oder modifizierten Parametern. Dies ermöglicht eine effizientere Anpassung an spezifische Aufgaben oder Domänen, ohne die Notwendigkeit enormer Hardware-Ressourcen.



<p><font color='black' size="5">
LoRA (Low-Rank Adaptation)
</font></p>

LoRA (Low-Rank Adaptation) ist eine populäre PEFT-Methode, die die Feinabstimmung von LLMs optimiert, indem sie nur eine begrenzte Anzahl von Parametern anpasst. Die Kernidee von LoRA besteht darin, trainierbare Low-Rank-Matrizen zu den ursprünglichen Gewichtsmatrizen des Basismodells hinzuzufügen. Anstatt alle Parameter zu modifizieren, werden lediglich zwei kleinere Matrizen trainiert, die dann mit den Originalgewichten kombiniert werden. Typischerweise werden bei LoRA nur die Attention-Layer oder lineare Schichten des Modells angepasst, was lediglich 1-10 % der gesamten Parameter ausmacht.

Die Vorteile von LoRA sind vielfältig:
* **Ressourceneffizienz:** Es reduziert den GPU-Speicherbedarf erheblich (bis zu 33 % weniger als Full Fine-Tuning).
* **Overfitting-Vermeidung:** Die geringere Anzahl an trainierbaren Parametern minimiert das Risiko, dass das Modell irrelevante Muster aus den Trainingsdaten lernt.
* **Flexibilität:** Die trainierten LoRA-Adapter können modular für verschiedene Aufgaben hinzugefügt oder entfernt werden.




<p><font color='black' size="5">
QLoRA (Quantized LoRA)
</font></p>

QLoRA (Quantized LoRA) baut auf LoRA auf, indem es zusätzlich Quantisierungstechniken einsetzt, um Modelle auf eine geringere Bit-Genauigkeit (z. B. 4-8 Bit) zu komprimieren. **Quantisierung** ist ein Prozess, bei dem hochpräzise Werte (z.B. 32-Bit-Fließkommazahlen) in Werte mit geringerer Genauigkeit (z.B. 8-Bit- oder 4-Bit-Ganzzahlen) umgewandelt werden. Ziel ist es, Speicherplatz zu sparen und die Verarbeitungsgeschwindigkeit zu erhöhen.

Die Kombination von LoRA und Quantisierung ermöglicht es, sehr große Modelle (z.B. mit 30 Milliarden Parametern) auf Consumer-GPUs zu trainieren und eine nahezu verlustfreie Rekonstruktion der Originalgewichte nach dem Training zu erreichen.




<p><font color='black' size="5">
Vergleich zu Full Fine-Tuning
</font></p>

Der Vergleich zwischen LoRA/QLoRA und Full Fine-Tuning zeigt die erheblichen Vorteile der PEFT-Methoden:

| Aspekt              | LoRA/QLoRA | Full Finetuning |
| :------------------ | :--------- | :-------------- |
| Trainingsparameter  | 1-10 %     | 100 %           |
| GPU-Speicher        | 8-16 GB    | 40-80 GB        |
| Overfitting-Risiko  | Niedrig    | Hoch            |
| Trainingsgeschwindigkeit | Schnell    | Langsam         |

LoRA und QLoRA bieten somit einen exzellenten Kompromiss zwischen Leistung und Effizienz, was sie besonders wertvoll für Szenarien mit begrenzten Rechenressourcen macht. Es ist jedoch wichtig zu beachten, dass OpenAI derzeit keine direkte Unterstützung für LoRA oder QLoRA beim Fine-Tuning seiner Close-Source-Modelle wie GPT-4 anbietet. Wer diese spezifischen Techniken nutzen möchte, greift typischerweise auf Open-Source-Modelle aus der Hugging Face-Community zurück.

````python
import re
import warnings
import json
import torch

from datasets import Dataset
from unsloth import FastLanguageModel

MAX_SEQ_LENGTH = 2048
NUM_TRAIN_EPOCHS = 5
MAX_NEW_TOKENS = 256
REPETITION_PENALTY = 1.2
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
GGUF_EXPORT_BASENAME = "gguf_model_scratch_fixed"
GGUF_EXPORT_DIR = f"{GGUF_EXPORT_BASENAME}_gguf"

torch.manual_seed(42)
print(f"Device: {DEVICE}")
````

## 3.3 Basismodell laden

Hier wird `unsloth/Qwen3-4B-Instruct-2507-unsloth-bnb-4bit` geladen. Die 4-bit-Variante spart GPU-Speicher und macht das Fine-Tuning auch in begrenzten Notebook-Umgebungen praktikabler.

`MAX_SEQ_LENGTH` begrenzt die nutzbare Kontextlänge während Training und Inferenz. Für kurze Frage-Antwort-Paare reicht das aus; bei längeren Dokumenten oder mehrteiligen Dialogen müsste dieser Wert bewusst geprüft werden.

````python
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Qwen3-4B-Instruct-2507-unsloth-bnb-4bit",
    max_seq_length = MAX_SEQ_LENGTH,
    dtype = None,
    load_in_4bit = True
)
````

## 3.4 Trainingsdaten vorbereiten

Die Rohdaten liegen bereits im `messages`-Format vor und werden anschließend in das Chat-Template des Tokenizers übertragen. Dadurch sieht jedes Trainingsbeispiel so aus, wie das Modell auch später in der Inferenz angesprochen wird: eine Nutzerrolle mit Frage und eine Assistentenrolle mit Zielantwort.

Ein einzelner Datensatz-Eintrag hat diese Struktur:

```json
{
  "messages": [
    {"role": "user", "content": "..."},
    {"role": "assistant", "content": "..."}
  ]
}
```

Dieser Schritt ist wichtiger als er aussieht. Wenn Trainingsformat und Inferenzformat nicht zusammenpassen, kann das Modell trotz erfolgreichem Training schlechter antworten.

````python
import os

_filename = "pneumatische_plastologie_trainset_en_clean_reinforced.json"
_colab   = f"/content/{_filename}"
_local   = _filename  # falls Datei im Arbeitsverzeichnis liegt

if os.path.exists(_colab):
    _path = _colab
elif os.path.exists(_local):
    _path = _local
else:
    raise FileNotFoundError(f"Trainset nicht gefunden: weder {_colab} noch {_local}")

with open(_path, "r", encoding="utf-8") as f:
    data = json.load(f)

raw_ds = Dataset.from_list(data)
raw_split = raw_ds.train_test_split(test_size=0.1, seed=42)
train_examples = raw_split["train"]
val_examples = raw_split["test"]

def to_text(ex):
    return {
        "text": tokenizer.apply_chat_template(
            ex["messages"],
            tokenize=False,
            add_generation_prompt=False,
        )
    }

train_ds = train_examples.map(to_text, remove_columns=train_examples.column_names)
val_ds = val_examples.map(to_text, remove_columns=val_examples.column_names)

print(f"Train: {len(train_ds)} | Val: {len(val_ds)}")
````

## 3.5 LoRA-Adapter konfigurieren

Beim LoRA-Fine-Tuning werden nicht alle Modellgewichte verändert. Stattdessen lernt das Notebook zusätzliche kleine Adaptermatrizen in ausgewählten Schichten. Das reduziert Speicherbedarf und Trainingszeit deutlich.

`r = 64` bestimmt den Rang der Adapter und damit ihre Kapazität. `lora_alpha = 128` skaliert den Einfluss der Adapter. Die Zielmodule umfassen Attention- und MLP-Projektionen, wodurch das Modell sowohl Antwortstruktur als auch domänenspezifische Inhalte anpassen kann.

````python
model = FastLanguageModel.get_peft_model(
    model,
    r = 64,  # rank of matrices (for LoRA)
    target_modules=[
        'q_proj', 'k_proj', 'v_proj', 'o_proj',
        'gate_proj', 'up_proj', 'down_proj',
    ],
    lora_alpha = 64 * 2,  # scaling factor, usually 2x rank
    lora_dropout = 0,
    bias = 'none',
    use_gradient_checkpointing = 'unsloth',
)
````

**Was passiert hier?**

1. `FastLanguageModel.get_peft_model(...)` ergänzt LoRA-Adapter zum geladenen Basismodell.
2. `target_modules` legt fest, welche Attention- und MLP-Projektionen trainierbare Adapter erhalten.
3. `r`, `lora_alpha` und `lora_dropout` steuern Kapazität, Skalierung und Regularisierung der Adapter.
4. `use_gradient_checkpointing="unsloth"` reduziert den Speicherbedarf während des Trainings.

## 3.6 Supervised Fine-Tuning starten

Der `SFTTrainer` trainiert das Modell darauf, zu jedem Prompt die jeweilige Zielantwort zu erzeugen. Die effektive Batch-Größe ergibt sich aus `per_device_train_batch_size` mal `gradient_accumulation_steps`.

Mit dem erweiterten, konsistenteren Datensatz werden wieder 5 Epochen verwendet. Der Trainingsverlust sollte sinken, ist aber keine Qualitätsgarantie. Entscheidend sind die anschließenden deterministischen Testfragen mit erwarteten Antworten.

````python
from trl import SFTTrainer, SFTConfig
from transformers import TrainingArguments

trainer = SFTTrainer(  # supervised fine-tuning trainer
    model = model,
    train_dataset = train_ds,
    eval_dataset  = val_ds,
    tokenizer = tokenizer,
    dataset_text_field = 'text',
    max_seq_length = MAX_SEQ_LENGTH,
    args = SFTConfig(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 10,
        max_steps = -1,
        logging_steps = 1,
        eval_strategy = "epoch",
        output_dir = "outputs",
        optim = "adamw_8bit",
        num_train_epochs = NUM_TRAIN_EPOCHS,
        seed = 42,
    ),
)

trainer.train()
````

**Was passiert hier?**

1. `SFTTrainer` erhält Modell, Tokenizer, Trainingsdaten und Validierungsdaten.
2. `dataset_text_field="text"` sagt dem Trainer, dass die vorbereiteten Chat-Template-Texte trainiert werden.
3. `SFTConfig` bündelt Batch-Größe, Akkumulation, Epochenzahl, Logging und Evaluation.
4. `trainer.train()` startet das supervised Fine-Tuning und schreibt Checkpoints nach `outputs`.

## 4.2 Deterministische Testfunktion

Die Funktion `ask_llm` verpackt eine Nutzerfrage im passenden Chat-Template, übergibt sie an das Modell und dekodiert nur die neu erzeugten Tokens.

Für die Qualitätsprüfung wird **greedy Decoding** verwendet: `do_sample=False`. Damit sind die Testantworten deterministisch und besser mit den erwarteten Antworten vergleichbar. Wenn diese deterministischen Antworten falsch sind, liegt das Problem eher bei Training, Datenmenge, Datenformat oder Modellanpassung als bei zufälligem Sampling.

````python
def ask_llm(user_prompt: str) -> str:
    messages = [{"role": "user", "content": user_prompt}]

    inputs = tokenizer.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        return_tensors="pt",
    ).to(DEVICE)

    outputs = model.generate(
        inputs,
        max_new_tokens=MAX_NEW_TOKENS,
        do_sample=False,
        repetition_penalty=REPETITION_PENALTY,
        pad_token_id=tokenizer.pad_token_id,
    )

    result = tokenizer.decode(
        outputs[0][inputs.shape[1]:],
        skip_special_tokens=True,
    )

    return result.replace("<tool_call>", "").replace("</tool_call>", "").strip()
````

## 4.3 Zufällige Beispiele und JUDGE-Bewertung

Für die Qualitätsprüfung werden bei jedem Lauf fünf zufällige Frage-Antwort-Paare aus dem Validierungsbestand (`val_examples`) ausgewählt. Das fine-tuned Modell beantwortet diese Fragen erneut. Anschließend bewertet ein GPT-Modell (`JUDGE`), wie gut die Modellantworten zu den erwarteten Antworten passen, und fasst die Ergebnisse zusammen.

Diese Prüfung nutzt den 10%-Validierungssplit und ist damit aussagekräftiger als eine Stichprobe aus den Trainingsbeispielen. Sie ersetzt aber keine systematische Evaluation mit einem vorher getrennten Holdout-Set.

> [!NOTE] Hinweis<br>
> Dieses Beispiel zeigt den Workflow und die iterative Verbesserung. Für eine belastbare Qualitätsaussage wäre ein getrenntes Holdout-Set nötig.

> **Hinweis:** Fine-Tuning ist oft ein iterativer Prozess. Nach einem Trainingslauf werden typische Fehler analysiert, der Datensatz gezielt korrigiert oder ergänzt und anschließend erneut trainiert. Diese schrittweise Verfeinerung ist normal. Sie sollte aber nicht nur auf zufällige Einzelbeispiele optimieren, sondern mit festen Testfragen oder einem getrennten Holdout-Set überprüft werden.

````python
import random
from langchain.chat_models import init_chat_model

examples = random.sample(list(val_examples), k=min(5, len(val_examples)))
evaluation_items = []

for i, ex in enumerate(examples, start=1):
    question = ex["messages"][0]["content"]
    expected = ex["messages"][1]["content"]
    actual = ask_llm(question)

    evaluation_items.append({
        "question": question,
        "expected": expected,
        "actual": actual,
    })

    mprint(f"### Beispiel {i}")
    print("Question:")
    print(question)
    print()
    print("Expected answer:")
    print(expected)
    print()
    print("Model answer:")
    print(actual)
    mprint("\n" + "---" "\n")

judge_prompt = json.dumps(evaluation_items, ensure_ascii=False, indent=2)

judge_llm = init_chat_model(JUDGE)
judge_response = judge_llm.invoke([
    (
        "system",
        "Du bist ein strenger, aber pragmatischer Evaluator für Fine-Tuning-Ergebnisse. "
        "Vergleiche jede Modellantwort mit der erwarteten Antwort. "
        "Bewerte jedes Beispiel mit 0, 1 oder 2 Punkten: "
        "0 = falsch oder Halluzination, 1 = teilweise korrekt, 2 = korrekt. "
        "Nenne anschließend Gesamtscore, wichtigste Schwächen und eine kurze Einordnung. "
        "Die Einordnung muss zusätzlich als Markdown-Tabelle mit den Spalten Kategorie und Befund ausgegeben werden. "
        "Der Befund ist kurz und knapp zu formulieren. "
        "Verwende mindestens diese Kategorien: Domänenwissen, Faktentreue bei seltenen Details, Instruktionsbefolgung, Rollenzuordnung."
    ),
    (
        "human",
        "Bewerte diese Fine-Tuning-Ergebnisse:\n\n" + judge_prompt
    ),
])

mprint("## Bewertung durch JUDGE")
mprint(judge_response.content)
````

**Was passiert hier?**

1. `random.sample(list(val_examples), k=min(5, len(val_examples)))` zieht bei jedem Lauf bis zu fünf Beispiele aus dem Validierungsbestand.
2. `ask_llm(...)` lässt das fine-tuned Modell jede Frage deterministisch beantworten.
3. Die erwartete Antwort und die Modellantwort werden als JSON an `JUDGE` übergeben.
4. `JUDGE` bewertet die Antwortqualität und fasst typische Fehler zusammen.

## 5.2 GGUF-Dateien prüfen und Nutzung anzeigen

Wenn der Export ausgeführt wurde, prüft die nächste Zelle, ob eine GGUF-Datei und ein `Modelfile` entstanden sind. Danach werden die wichtigsten Aufrufmuster für llama.cpp und Ollama ausgegeben.

````python
if EXPORT_MODEL:
    import os

    export_dir = GGUF_EXPORT_DIR
    expected_files = [
        os.path.join(export_dir, "Modelfile"),
    ]

    missing_files = [path for path in expected_files if not os.path.exists(path)]
    gguf_files = [name for name in os.listdir(export_dir) if name.endswith(".gguf")] if os.path.isdir(export_dir) else []

    assert not missing_files, f"Fehlende Exportdateien: {missing_files}"
    assert gguf_files, "Keine GGUF-Datei im Exportverzeichnis gefunden."

    # ── Nutzung des exportierten Modells ──────────────────────────────────────
    print("Export-Ergebnis:")
    print(f"  → {export_dir}/{gguf_files[0]}")
    print(f"  → {export_dir}/Modelfile")
    print()
    print("llama.cpp:  /root/.unsloth/llama.cpp/llama-cli -m <gguf-datei> -p '<prompt>' -n 256")
    print("Ollama:     ollama create <name> -f <Modelfile> && ollama run <name>")
else:
    print("Exportprüfung übersprungen, weil kein Export ausgeführt wurde.")
````
