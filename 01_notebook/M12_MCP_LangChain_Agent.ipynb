# 1 | Intro

---

<p><font color='black' size="5">
Was ist MCP (Model Context Protocol)?
</font></p>

MCP ist ein **standardisiertes Protokoll**, das LLMs (Large Language Models) mit externen Tools und Datenquellen verbindet. MCP ist wie ein **USB fГјr AI** - ein universeller Standard fГјr Tool-Integration.

**Warum MCP?**
- вң… **Standardisiert**: Ein Protokoll fГјr alle Tools
- вң… **Wiederverwendbar**: Einmal erstellen, Гјberall nutzen
- вң… **Skalierbar**: Einfach neue Tools hinzufГјgen
- вң… **Production-Ready**: FГјr echte Anwendungen gedacht

**Architektur:**
```
LangChain Agent
        вҶ“
MCP Client (langchain-mcp-adapters)
        вҶ“ HTTP POST (JSON-RPC 2.0)
MCP Server (FastMCP auf localhost:8000)
        вҶ“
Tools (add, multiply, subtract, divide)
```

<p><font color='black' size="5">
Wie funktioniert MCP?
</font></p>



MCP basiert auf einem **Client-Server-Modell** mit drei Hauptkomponenten:

**1. MCP Server**
- Stellt **Tools** (Funktionen) bereit
- Kann **Ressourcen** (Daten) anbieten
- Implementiert das MCP-Protokoll (JSON-RPC 2.0)
- LГӨuft als eigenstГӨndiger Prozess

**2. MCP Client**
- Verbindet sich mit einem oder mehreren Servern
- LГӨdt verfГјgbare Tools vom Server
- Гңbersetzt LLM-Anfragen in MCP-Protokoll
- Verwaltet die Kommunikation

**3. LLM/Agent**
- Entscheidet, welche Tools benГ¶tigt werden
- Orchestriert mehrere Tool-Aufrufe
- Generiert natГјrlichsprachliche Antworten
- Nutzt Tools Гјber den Client

**Transport:**

- `streamable_http`: HTTP-basiert (fГјr Web-Services) вң… In diesem Notebook


вңҸпёҸ Vergleich der MCP Transport-Optionen

<details>



| Transport | Beschreibung | Funktionsweise | PrimГӨrer Use-Case |
| --- | --- | --- | --- |
| **`stdio`** | **Standard Input/Output** | Daten werden Гјber die Konsole (`stdin`/`stdout`) ausgetauscht. | **Lokale Integration:** Wenn der LLM-Client (z.B. Claude Desktop) den Server direkt als Prozess startet. |
| **`streamable_http`** | **HTTP-basiert** | Nutzt Standard `HTTP POST` Anfragen fГјr den Austausch von JSON-RPC Objekten. | **Web-Services:** Ideal fГјr Cloud-Infrastrukturen und Notebook-Umgebungen (wie in diesem Beispiel). |
| **`sse`** | **Server-Sent Events** | Ein unidirektionaler Stream vom Server zum Client Гјber eine offene HTTP-Verbindung. | **Streaming:** Wenn der Server Antworten in Echtzeit "pushen" muss, ohne auf neue Anfragen zu warten. |
| **`websocket`** | **WebSockets** | Eine dauerhafte, voll-bidirektionale Verbindung zwischen beiden Parteien. | **Hochperformante Apps:** Wenn minimale Latenz und stГӨndiger Datenaustausch in beide Richtungen nГ¶tig sind. |

</details>


вңҸпёҸ MCP-Infrastruktur: Von der Verbindung zum Befehl

<details>

   

1. Das Fundament: **HTTP (Transport)**
Bevor Daten flieГҹen, wird eine Verbindung via HTTP (Hypertext Transfer Protocol) hergestellt.

    + Der Server als Web-Service: Erreichbar unter einer festen Adresse (URL).

    + Die Methode: Nutzt HTTP POST, um Datenpakete gezielt an den Server zu senden.

    + Vorteil: Funktioniert Гјberall (Firewall-freundlich) und ist der Standard fГјr moderne Software-Kommunikation.

2. Der Inhalt: **JSON-RPC 2.0 (Protokoll)**
Innerhalb des HTTP-Transports wird die "Sprache" JSON-RPC gesprochen. Sie strukturiert den eigentlichen Befehl.

    + Anfrage: EnthГӨlt die id (Zuordnung), die method (Tool-Name) und params (Eingabewerte).

    + Antwort: Liefert das result oder eine Fehlermeldung zurГјck an den Agenten.

    + Vorteil: Schlank und prГӨzise вҖ“ es wird nur das Гјbertragen, was fГјr den Funktionsaufruf nГ¶tig ist.

3. KurzГјbersicht: Die Hierarchie        
    + Transport-Layer (HTTP): Der LKW, der das Paket von A nach B bringt.
    + Payload-Layer (JSON-RPC): Der Inhalt des Pakets (der eigentliche Arbeitsauftrag).

Fazit: Erst wenn der HTTP-Kanal steht, kann JSON-RPC die Tools des MCP-Servers steuern.


</details>

вңҸпёҸ Model Context Protocol (MCP) вҖ“ JSON-RPC Kommunikation

<details>

Das MCP nutzt **JSON-RPC 2.0 Гјber HTTP**. In der Praxis bedeutet das:

1. Der **Client** sendet eine `HTTP POST`-Anfrage an den Server.
2. Der **Body** enthГӨlt ein JSON-Objekt nach der JSON-RPC 2.0 Spezifikation.
3. Der **Server** antwortet mit `HTTP 200 OK` und dem entsprechenden Ergebnis im Body.

---

**Beispiel: Tool-Aufruf (`add`)**

Stellen wir uns vor, ein LangChain Agent mГ¶chte das Tool `add` mit den Werten  und  aufrufen.

*1. Client an Server (Request)*

Der Client sendet eine Anfrage an `http://127.0.0.1:8000/mcp`:

```json
{
  "jsonrpc": "2.0",
  "method": "add",
  "params": {
    "a": 3,
    "b": 5
  },
  "id": "some-unique-request-id"
}

```

ErklГӨrung der Felder:

* **`jsonrpc`**: Die Protokoll-Version (immer `"2.0"`).
* **`method`**: Der Name des aufzurufenden Tools (hier: `add`).
* **`params`**: Ein Objekt mit den Argumenten als SchlГјssel-Wert-Paare (entspricht den Python-Parametern `a` und `b`).
* **`id`**: Eine eindeutige ID, um die Antwort spГӨter der richtigen Anfrage zuzuordnen.

---

*2. Server an Client (Response)*

Nach erfolgreicher AusfГјhrung sendet der Server das Ergebnis zurГјck:

```json
{
  "jsonrpc": "2.0",
  "result": 8,
  "id": "some-unique-request-id"
}

```

* **`result`**: Der berechnete RГјckgabewert (hier: ).
* **`id`**: Muss exakt mit der ID der Anfrage Гјbereinstimmen.

---

Fehlerbehandlung:

Sollte wГӨhrend der AusfГјhrung ein Fehler auftreten (z. B. eine Division durch Null), liefert der Server ein `error`-Objekt anstelle des `result`-Feldes:

```json
{
  "jsonrpc": "2.0",
  "error": {
    "code": -32000,
    "message": "ValueError: Division durch 0 ist nicht erlaubt"
  },
  "id": "some-unique-request-id"
}

```

---

> Dies ist die Datenstruktur, die Гјber die `streamable_http`-Verbindung ausgetauscht wird. Tools wie der `langchain-mcp-adapters` Client und der `FastMCP` Server Гјbernehmen die Serialisierung automatisch.

</details>

# 2 | FastMCP HTTP-Server starten

---

**FastMCP** ist ein Python-Framework, das die Erstellung von MCP-Servern (Model Context Protocol) radikal vereinfacht. Der Name ist angelehnt an FastAPI вҖ“ die gleiche Philosophie: minimaler Boilerplate-Code (repetitiver, standardisierter Code, der fГјr die Grundstruktur eines Programms nГ¶tig ist, aber keine eigentliche GeschГӨftslogik enthГӨlt), maximale FunktionalitГӨt.

**Kernidee**

```python
from fastmcp import FastMCP

mcp = FastMCP("MeinServer")

@mcp.tool()
def add(a: int, b: int) -> int:
    """Addiert zwei Zahlen."""
    return a + b
```

Mit nur wenigen Zeilen wird eine Python-Funktion zu einem Tool, das jedes LLM Гјber das MCP-Protokoll aufrufen kann.

**Hauptmerkmale**

| Feature | Beschreibung |
|---------|--------------|
| **Decorator-basiert** | `@mcp.tool()` macht jede Funktion zum MCP-Tool |
| **Automatische Schemas** | Type Hints werden zu JSON-Schemas fГјr das LLM |
| **Mehrere Transports** | HTTP, stdio, SSE, WebSocket |
| **Docstrings вҶ’ Beschreibungen** | Das LLM "liest" die Docstrings, um Tools zu verstehen |

**Warum FastMCP?**

Ohne FastMCP mГјsste man das JSON-RPC 2.0 Protokoll, Session-Management und Tool-Schemas manuell implementieren. FastMCP abstrahiert das alles вҖ“ man schreibt nur die eigentliche Logik.

**Kurz:** FastMCP ist fГјr MCP das, was FastAPI fГјr REST-APIs ist вҖ“ der schnellste Weg vom Python-Code zum produktionsreifen Server.

<p><font color='black' size="5">
Server erstellen
</font></p>

Es wird ein FastMCP-Server *erstellt*, der dann als **subprocess** gestartet wird.

````python
#@title рҹҡҖ MCP HTTP-Server erstellen und starten{ display-mode: "form" }

import subprocess
import time
import sys
import os

mprint("### рҹҡҖ MCP HTTP-Server erstellen")
mprint("---")

# SCHRITT 1: Server-Datei erstellen
server_code = '''
"""FastMCP Math Server - HTTP Transport"""
from fastmcp import FastMCP

# FastMCP Server initialisieren
mcp = FastMCP("MathTools")

@mcp.tool()
def add(a: int, b: int) -> int:
    """Addiert zwei Zahlen."""
    return a + b

@mcp.tool()
def multiply(a: int, b: int) -> int:
    """Multipliziert zwei Zahlen."""
    return a * b

@mcp.tool()
def subtract(a: int, b: int) -> int:
    """Subtrahiert b von a."""
    return a - b

@mcp.tool()
def divide(a: float, b: float) -> float:
    """Teilt a durch b."""
    if b == 0:
        raise ValueError("Division durch 0 ist nicht erlaubt")
    return a / b

if __name__ == "__main__":
    # Server mit HTTP Transport starten
    mcp.run(transport="http", host="127.0.0.1", port=8000, path="/mcp")
'''

# Server-Datei schreiben
with open('math_mcp_server.py', 'w', encoding='utf-8') as f:
    f.write(server_code)

print("вң… Server-Datei erstellt: math_mcp_server.py")
print("   вҖў 4 Tools: add, multiply, subtract, divide\n")

# SCHRITT 2: Server als subprocess starten
print("вҸі Starte Server als subprocess...")

# PrГјfe ob Server bereits lГӨuft
try:
    import requests
    requests.get("http://127.0.0.1:8000", timeout=1)
    print("вҡ пёҸ Server lГӨuft bereits auf Port 8000")
    print("   Bitte fГјhre die Cleanup-Zelle aus oder starte den Kernel neu.\n")
except:
    pass  # Server lГӨuft nicht, gut!

# Server starten
server_process = subprocess.Popen(
    [sys.executable, "math_mcp_server.py"],
    stdout=subprocess.DEVNULL,
    stderr=subprocess.DEVNULL
)

# Warte bis Server bereit ist
print("вҸі Warte auf Server-Start...")
max_retries = 10
for i in range(max_retries):
    try:
        import requests
        response = requests.post(
            "http://127.0.0.1:8000/mcp",
            json={"jsonrpc": "2.0", "method": "tools/list", "id": 1},
            timeout=2
        )
        if response.status_code in [200, 400, 406]:
            # Server antwortet (auch Fehler = Server lГӨuft)
            break
    except:
        time.sleep(1)
        if i == max_retries - 1:
            print("вқҢ Server konnte nicht gestartet werden!")
            server_process.kill()
            raise Exception("Server-Start fehlgeschlagen")

mprint("\nвң… MCP-Server lГӨuft auf http://127.0.0.1:8000/mcp")
print(f"рҹ’Ў Server-Prozess PID: {server_process.pid}")
print("рҹ’Ў Der Server lГӨuft als echter subprocess!")
````

<details>

<summary><strong>
Ports & localhost
</strong></summary

</details>


Beim Aufruf einer URL wie `http://127.0.0.1:8000/1/mcp` lГӨuft intern folgender Prozess ab:

Der Rechner kommuniziert mit sich selbst (Гјber `127.0.0.1`) und leitet die Anfrage an ein **konkretes Programm** weiter вҖ“ identifiziert Гјber den **Port**.

Da mehrere Programme gleichzeitig Netzwerkverbindungen nutzen (z. B. Browser, Datenbank, API), benГ¶tigt jedes Programm einen eigenen вҖһEingangвҖң.
Dieser Eingang wird als **Port** bezeichnet.

---

**Kurzdefinitionen**

**`127.0.0.1` (localhost)**    
вҶ’ Eigener Rechner (Loopback)   
вҶ’ Kommunikation bleibt vollstГӨndig lokal   

**Ports (0вҖ“65535)**    
вҶ’ Virtuelle EingГӨnge fГјr Programme    
вҶ’ ErmГ¶glichen parallele Netzwerkkommunikation    

---

**Komplett-Beispiel (MCP)**

```id="k2y7vd"
http://127.0.0.1:8000/1/mcp
        в””в”Җв”Җв”Җв”Җв”Җв”Җв”¬в”Җв”Җв”Җв”Җв”Җв”Җв”ҳ
               в”Ӯ
     Adresse + Port = Zielprogramm
```

| Komponente  | Bedeutung                                 | Beispiel     |
| ----------- | ----------------------------------------- | ------------ |
| `127.0.0.1` | Lokaler Rechner (nicht extern erreichbar) | Dev-Server   |
| `:8000`     | Port вҶ’ welches Programm hГ¶rt dort?        | FastAPI/MCP  |
| `/1/mcp`    | Route innerhalb der Anwendung             | API-Endpunkt |

---

**Wichtige PrГӨzisierung**

* Ein **Port ist kein Programm**
* Ein Programm **bindet sich an einen Port**
* Beispiel:    
  вҶ’ FastAPI auf Port 8000   
  вҶ’ anderes Tool z. B. auf Port 3000    

---

**Mentales Modell**

* IP = **Hausadresse**
* Port = **Wohnung**
* Pfad = **Person in der Wohnung**

---

**Merksatz**

> `127.0.0.1` = Kommunikation mit dem eigenen Rechner     
> `:8000` = Weiterleitung an die Anwendung, die dort lauscht   

---

**Kritischer Zusatz**

вҖһLokalвҖң bedeutet lediglich: **nicht von auГҹen erreichbar**      
вҶ’ keine Garantie fГјr Sicherheit (z. B. bei unsicheren APIs)

Server-Status testen

Man kann den Server direkt Гјber HTTP testen:

````python
#@title рҹ§Ә Server-Health-Check{ display-mode: "form" }
import requests

try:
    mprint("### рҹ§Ә Server Health Check (Status)")
    mprint("---")
except NameError:
    print("### рҹ§Ә Server Health Check (Status)")
    print("---")

try:
    # Wir senden einen einfachen Request.
    # Wir erwarten fast schon einen Fehler 400 oder 406,
    # weil wir keine echte Session aufbauen.
    response = requests.get(
        "http://127.0.0.1:8000/mcp",
        timeout=5
    )

    # PrГјfung: Lebt der Server?
    # Status 400 (Bad Request) mit MCP-Fehlermeldung ist hier ein GUTES Zeichen!
    # Es bedeutet: Der Server lГӨuft, wartet aber auf einen korrekten Client.

    server_is_alive = False

    if response.status_code == 200:
        server_is_alive = True
    elif response.status_code in [400, 405, 406]:
        # PrГјfen, ob die Antwort JSON ist (typisch fГјr MCP)
        if "application/json" in response.headers.get("Content-Type", ""):
            server_is_alive = True

    if server_is_alive:
        print(f"вң… MCP-Server ist ONLINE und erreichbar!")
        print(f"   Status: {response.status_code} (Erwartet)")
        print(f"   Antwort: Der Server hat geantwortet (MCP-Logik lГӨuft).")
        print(f"рҹҢҗ URL: http://127.0.0.1:8000/mcp")
        print("\nрҹ’Ў Man kann jetzt mit dem nГӨchsten Schritt (LangChain Client) fortfahren.")
        print("   Der Client Гјbernimmt das Session-Management automatisch.")
    else:
        print(f"вҡ пёҸ Unerwartete Antwort: Status {response.status_code}")
        print(f"   Text: {response.text}")

except requests.exceptions.ConnectionError:
    print("вқҢ Server nicht erreichbar (Connection Refused).")
    print("   Bitte starten Sie die Zelle 'MCP HTTP-Server erstellen' neu.")
except Exception as e:
    print(f"вқҢ Fehler: {e}")
````

# 3 | LangChain MCP Client

Der **MCPClient** verbindet sich Гјber **streamable_http** mit dem Server:

````python
#@title рҹ§ңвҖҚвҷҖпёҸ Sequenz-Diagramm { display-mode: "form" }

diagram = '''
sequenceDiagram

    autonumber
    actor User
    participant Agent as LangChain Agent
    participant Client as MCP Client
    participant Server as MCP Server (FastMCP)

    Note over Agent, Server: Initialisierung (Handshake)

    Client->>Server: HTTP POST (method: initialize)
    Server-->>Client: JSON Response (Capabilities & Version)
    Client->>Server: HTTP POST (notifications/initialized)


    Note over Agent, Server: Tools abfragen (await client.get_tools())


    Client->>Server: HTTP POST (method: tools/list)
    Note right of Client: JSON-RPC 2.0 Request

    Server-->>Client: JSON Response (Liste der Tool-Definitionen)
    Note left of Server: Name, Description, Input Schema



'''
mermaid(diagram, width=800, height=550)
````

````python
#@title рҹ”Ң MCP Client mit HTTP initialisieren{ display-mode: "form" }

from langchain_mcp_adapters.client import MultiServerMCPClient

mprint("### рҹ”Ң MCP Client Setup (HTTP)")
mprint("---")

# Client konfigurieren mit streamable_http Transport (nicht "http"!)
client = MultiServerMCPClient(
    {
        "math": {
            "transport": "streamable_http",  # вң… Richtiger Transport-Name!
            "url": "http://127.0.0.1:8000/mcp"  # HTTP-Endpoint
        }
    }
)

print("вң… Client konfiguriert")
print(f"   Transport: streamable_http")
print(f"   Server-URL: http://127.0.0.1:8000/mcp")
````

# 4 | LangChain Agent

<p><font color='black' size="5">
Agent mit MCP-Tools erstellen

вңҸпёҸ Details: Model Context Protocol (MCP) вҖ“ JSON-RPC Kommunikation

<details>


Das MCP nutzt **JSON-RPC 2.0 Гјber HTTP**. In der Praxis bedeutet das:

1. Der **Client** sendet eine `HTTP POST`-Anfrage an den Server.
2. Der **Body** enthГӨlt ein JSON-Objekt nach der JSON-RPC 2.0 Spezifikation.
3. Der **Server** antwortet mit `HTTP 200 OK` und dem entsprechenden Ergebnis im Body.

---

**Beispiel: Tool-Aufruf (`add`)**

Stellen wir uns vor, ein LangChain Agent mГ¶chte das Tool `add` mit den Werten  und  aufrufen.

*1. Client an Server (Request)*

Der Client sendet eine Anfrage an `http://127.0.0.1:8000/mcp`:

```json
{
  "jsonrpc": "2.0",
  "method": "add",
  "params": {
    "a": 3,
    "b": 5
  },
  "id": "some-unique-request-id"
}

```

ErklГӨrung der Felder:

* **`jsonrpc`**: Die Protokoll-Version (immer `"2.0"`).
* **`method`**: Der Name des aufzurufenden Tools (hier: `add`).
* **`params`**: Ein Objekt mit den Argumenten als SchlГјssel-Wert-Paare (entspricht den Python-Parametern `a` und `b`).
* **`id`**: Eine eindeutige ID, um die Antwort spГӨter der richtigen Anfrage zuzuordnen.

---

*2. Server an Client (Response)*

Nach erfolgreicher AusfГјhrung sendet der Server das Ergebnis zurГјck:

```json
{
  "jsonrpc": "2.0",
  "result": 8,
  "id": "some-unique-request-id"
}

```

* **`result`**: Der berechnete RГјckgabewert (hier: ).
* **`id`**: Muss exakt mit der ID der Anfrage Гјbereinstimmen.

---

Fehlerbehandlung:

Sollte wГӨhrend der AusfГјhrung ein Fehler auftreten (z. B. eine Division durch Null), liefert der Server ein `error`-Objekt anstelle des `result`-Feldes:

```json
{
  "jsonrpc": "2.0",
  "error": {
    "code": -32000,
    "message": "ValueError: Division durch 0 ist nicht erlaubt"
  },
  "id": "some-unique-request-id"
}

```

---

> Dies ist die Datenstruktur, die Гјber die `streamable_http`-Verbindung ausgetauscht wird. Tools wie der `langchain-mcp-adapters` Client und der `FastMCP` Server Гјbernehmen die Serialisierung automatisch.

</details>
