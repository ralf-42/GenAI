## DALE2 vs. DALE3

DALL·E 2 und DALL·E 3 sind aufeinanderfolgende Versionen der leistungsstarken Bildgenerierungsmodelle von OpenAI, die jeweils auf den Stärken und Fähigkeiten ihres Vorgängers aufbauen.

DALL·E 2, das 2022 auf den Markt kam, stellte einen bedeutenden Sprung gegenüber dem ursprünglichen DALL·E-Modell dar. Es war in der Lage, hochauflösende, hochdetaillierte Bilder aus Textaufforderungen zu erstellen, wobei der Schwerpunkt insbesondere auf kreativer Abstraktion und fantasievollen Kombinationen lag. Einige der Hauptfunktionen von DALL·E 2 sind:

* **Realismus und künstlerischer Ausdruck**: DALL·E 2 zeichnete sich durch die Generierung sowohl fotorealistischer Bilder als auch künstlerischer Interpretationen auf der Grundlage der Texteingabe aus. Es verfügte über eine bemerkenswerte Fähigkeit, verschiedene Elemente (z. B. surreale, abstrakte oder realistische Objekte) zu zusammenhängenden Kompositionen zu verschmelzen.
* **Inpainting**: Benutzer können bestimmte Teile eines Bildes ändern, indem sie die gewünschte Änderung beschreiben, was eine interaktivere Kontrolle über die Bilderstellung ermöglicht.
* **Stilvariationen**: DALL·E 2 kann verschiedene Kunststile wie Malerei oder Fotografie nachahmen und bietet so Flexibilität für kreative Projekte.

Obwohl DALL·E 2 sehr vielseitig war und beeindruckende Bilder erzeugte, ließen seine schnelle Auffassungsgabe und die Befolgung nuancierter Texte manchmal zu wünschen übrig. Es hatte Probleme mit komplexen oder mehrdeutigen Anweisungen und interpretierte gelegentlich subtile Details in der Benutzereingabe falsch.

DALL·E 3, das im Jahr 2023 veröffentlicht wurde, verbesserte die Einschränkungen von DALL·E 2 und führte neue Funktionen ein, die es zugänglicher und präziser machten:

* **Verbessertes Textverständnis**: Eines der herausragenden Merkmale von DALL·E 3 ist die verbesserte Fähigkeit, komplexere oder spezifischere Anweisungen zu verstehen und genau zu befolgen. Es erfasst Nuancen viel besser und erleichtert so die Erstellung von Bildern, die den Erwartungen des Benutzers entsprechen, selbst bei detaillierten oder komplizierten Anweisungen.
* **Tiefere Integration mit ChatGPT**: DALL·E 3 wurde tief in ChatGPT integriert, sodass Benutzer mit dem Modell in Dialogform interagieren und ihre Eingabeaufforderungen und Anfragen iterativ verfeinern können. Diese Integration vereinfacht den kreativen Prozess und ermöglicht intuitivere Anpassungen der Eingabeaufforderungen.
* **Reduzierung bekannter Einschränkungen**: DALL·E 3 hat die Handhabung schwieriger Aufgaben wie der Darstellung menschlicher Hände und Gesichter verbessert. In diesen Bereichen lieferten frühere Modelle häufig verzerrte oder unrealistische Ergebnisse.
Verbesserungen beim Inpainting: DALL·E 3 setzt die Entwicklung des Inpaintings fort und bietet noch mehr Kontrolle über Änderungen an bestimmten Bereichen der generierten Bilder. Dadurch können Benutzer Teile von Bildern verfeinern, ohne von vorne beginnen zu müssen.

Die Images API bietet drei Methoden zur Interaktion mit Bildern:

* Erstellen von Bildern von Grund auf basierend auf einer Textaufforderung (DALL·E 3 und DALL·E 2)
* Erstellen bearbeiteter Versionen von Bildern, indem das Modell einige Bereiche eines bereits vorhandenen Bildes ersetzt, basierend auf einer neuen Textaufforderung (nur DALL·E 2)
* Erstellen von Variationen eines vorhandenen Bildes (nur DALL·E 2)


## Mehrere Bilder generieren

Mit DALL·E 2 können Sie mehrere Bilder generieren, die wir in diesem Abschnitt sehen werden.

Der folgende Code generiert eine „weiße Siamkatze“.




````python
from openai import OpenAI
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

client = OpenAI()

# 3 Bilder generieren
response = client.images.generate(
    model="dall-e-2",
    prompt="a white siamese cat",
    size="1024x1024",
    quality="standard",
    n=3,  # 3 Bilder generieren
)

# Durchlaufen Sie die generierten Bilder, zeigen Sie sie mit Matplotlib an und speichern Sie sie als JPG
for i, data in enumerate(response.data):
    # Holen Sie sich die Bild-URL
    image_url = data.url

    # Holen Sie sich das Bild
    img_response = requests.get(image_url)
    img = Image.open(BytesIO(img_response.content))

    # Speichern Sie das Bild als JPG auf der Festplatte
    img.save(f"cat_image_{i+1}.jpg", "JPEG")

    # Zeigen Sie das Bild mit matplotlib an
    plt.imshow(img)
    plt.axis('off')  # Blenden Sie die Achsen aus, um eine übersichtlichere Ansicht zu erhalten
    plt.title(f"Image {i+1}")  # Fügen Sie einen Titel hinzu, um die Bilder zu unterscheiden
    plt.show()
````

# Teil 9.3: DALL·E Vorhandene Bilder

DALLE kann Bilder erweitern, indem es zusätzlichen Inhalt basierend auf dem Originalbild und einer bereitgestellten Eingabeaufforderung generiert. Diese Funktion, die oft als „Outpainting“ bezeichnet wird, ermöglicht es Benutzern, die Grenzen eines Bildes zu erweitern, Kontext hinzuzufügen oder eine Szene über den ursprünglichen Rahmen hinaus zu vervollständigen. Das Originalbild dient als Grundlage, während das Maskenbild verwendet wird, um die Bereiche zu definieren, in denen neuer Inhalt generiert wird. Die Maske, normalerweise ein transparenter oder ausgeblendeter Bereich, signalisiert DALLE, welche Teile der Leinwand ausgefüllt werden müssen. Durch die Angabe einer neuen Eingabeaufforderung weist der Benutzer DALLE an, wie der maskierte Bereich zu interpretieren ist, und bestimmt die Art der Erweiterung im Einklang mit dem Stil, Kontext oder Thema des Originalbilds.

In der Praxis nutzt das Übermalen das Originalbild und die Eingabeaufforderung, um die Kohärenz zwischen dem vorhandenen und dem generierten Inhalt aufrechtzuerhalten. Wenn das Originalbild beispielsweise eine Landschaft ist, kann die Eingabeaufforderung DALLE anleiten, diese Landschaft mit konsistenten Farben, Texturen und thematischen Elementen zu erweitern. Das System fügt die generierten Pixel nahtlos in das Original ein und erstellt so eine erweiterte Szene, die natürlich wirkt. Die Eingabeaufforderung spielt eine entscheidende Rolle bei der Steuerung des Generierungsprozesses und beschreibt die gewünschte Erweiterung in Bezug auf Objekte, Einstellungen oder kreative Richtungen, während die Maske sicherstellt, dass sich der neue Inhalt reibungslos in den vorgesehenen Bereich einfügt, ohne die vorhandenen Details zu überschreiben.

Bildquelle:
![Source Image](https://data.heatonresearch.com/images/wustl/app_genai/hickory_deck_1.png)

Maskenbild:
![Mask Image](https://data.heatonresearch.com/images/wustl/app_genai/hickory_deck_2.png)

Der folgende Code zeigt, wie das Bild erweitert wird.

````python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from openai import OpenAI

client = OpenAI()

# Funktion zum Herunterladen und Laden eines Bildes von einer URL
def load_image_from_url(url):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    return img

# URLs für die Bilder
image_url_1 = "https://data.heatonresearch.com/images/wustl/app_genai/hickory_deck_1.png"
image_url_2 = "https://data.heatonresearch.com/images/wustl/app_genai/hickory_deck_2.png"

# Laden Sie die Bilder von URLs
image_1 = load_image_from_url(image_url_1)
image_2 = load_image_from_url(image_url_2)

# Speichern Sie die Bilder in In-Memory-Dateien
img_byte_array_1 = BytesIO()
img_byte_array_2 = BytesIO()
image_1.save(img_byte_array_1, format='PNG')
image_2.save(img_byte_array_2, format='PNG')

# Setzt die Position des Puffers auf den Start zurück
img_byte_array_1.seek(0)
img_byte_array_2.seek(0)

# Verwenden Sie die OpenAI-API, um die Bilder mit In-Memory-Dateien zu bearbeiten
response = client.images.edit(
  image=img_byte_array_1,
  mask=img_byte_array_2,
  prompt="lion",
  n=2,
  size="1024x1024"
)

# Greifen Sie aus der Antwort auf die erste Bild-URL zu
edited_image_url = response.data[0].url

# Laden Sie das bearbeitete Bild von der URL herunter
edited_image_response = requests.get(edited_image_url)
resulting_image = Image.open(BytesIO(edited_image_response.content))

# Zeigen Sie das resultierende Bild mit matplotlib an
plt.imshow(resulting_image)
plt.axis('off')  # Die Achsen ausblenden
plt.show()
````

## Variationen erstellen

DALLE bietet die Möglichkeit, Bildvariationen zu erstellen. Mit dieser Funktion können Benutzer alternative Versionen eines vorhandenen Bildes mit geringfügigen Änderungen erstellen, während das Gesamtthema oder die Struktur beibehalten werden. Diese Funktion ist über die API in DALLE 2 verfügbar, wo Entwickler einen Bild-Input bereitstellen und darauf basierend mehrere Variationen anfordern können. Die API ermöglicht die Integration in Anwendungen, die die Erstellung kreativer Inhalte oder visuelles Brainstorming erfordern. Diese Bildvariationsfunktion wird in DALLE 3 jedoch noch nicht unterstützt, was bedeutet, dass Benutzer für diese spezielle Funktion weiterhin auf DALLE 2 angewiesen sind.

Wir werden uns ein Beispiel ansehen, das Variationen dieses Bildes erstellt.

![Bulldog Cartoon](https://data.heatonresearch.com/images/wustl/app_genai/hickory_cartoon.jpg)

````python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from openai import OpenAI

client = OpenAI()

# Funktion zum Anzeigen des Bildes mit matplotlib
def display_image(image):
    plt.imshow(image)
    plt.axis('off')  # Blenden Sie die Achsen aus, um eine übersichtlichere Ansicht zu erhalten
    plt.show()

# Funktion zum Herunterladen des Bildes von einer URL und Konvertieren in ein dateiähnliches Objekt
def load_image_from_url(url):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))

    # Speichern Sie das Bild in einer Datei im Arbeitsspeicher
    img_byte_array = BytesIO()
    img.save(img_byte_array, format='PNG')

    # Setzt die Position des Puffers auf den Start zurück
    img_byte_array.seek(0)
    return img_byte_array

# URL des zu verwendenden Bildes
image_url = "https://data.heatonresearch.com/images/wustl/app_genai/hickory_cartoon.jpg"

# Laden Sie das Bild von der URL
image_from_url = load_image_from_url(image_url)

# Erstellen Sie Variationen des Bildes mithilfe der OpenAI-API
response = client.images.create_variation(
    image=image_from_url,
    n=4,
    size="1024x1024"
)

# Durchlaufen der generierten Bildvarianten und Anzeigen/Speichern dieser
for i, data in enumerate(response.data):
    # Holen Sie sich die Bild-URL
    variation_image_url = data.url

    # Holen Sie sich das Bild von der URL
    variation_image_response = requests.get(variation_image_url)
    variation_image = Image.open(BytesIO(variation_image_response.content))

    # Zeigen Sie das Bild mit matplotlib an
    display_image(variation_image)

    # Optionales Speichern des Variationsbilds auf der Festplatte
    variation_image.save(f"variation_image_{i+1}.png", "PNG")
````

# Teil 9.4: Multimodale Modelle

Multimodale Modelle sind fortschrittliche KI-Systeme, die mehrere Arten von Eingaben wie Text, Bilder, Audio und Video verarbeiten und verstehen können. Diese Modelle kombinieren die Leistung verschiedener Datentypen, um komplexere Aufgaben auszuführen, die ein Verständnis der Beziehungen zwischen verschiedenen Modalitäten erfordern. Beispielsweise kann ein multimodales Modell ein Bild analysieren und beschreibenden Text generieren oder sowohl Text als auch Bilder zusammen verarbeiten, um differenziertere Erkenntnisse zu liefern. Durch die Integration verschiedener Modalitäten verbessern diese Modelle die Fähigkeit der KI, reale Daten zu interpretieren und mit ihnen zu interagieren, was sie für Anwendungen wie Bildbeschriftung, visuelle Fragenbeantwortung und mehr unverzichtbar macht.

In unserem Ansatz nutzen wir LangChain, ein Framework, das für die Entwicklung sprachmodellbasierter Anwendungen entwickelt wurde, um die Integration von Bildern in OpenAIs GPT-4o-mini zu ermöglichen. Durch die Verwendung von LangChain können wir Eingabeaufforderungen strukturieren, die Übertragung von Bildern handhaben und die Interaktion zwischen verschiedenen Modalitäten optimieren, wodurch GPT-4o-mini letztendlich zu einem vielseitigeren Tool für Aufgaben wird, die sowohl Bild- als auch Textverständnis erfordern. ## Bild zur Kunst


Mit diesem multimodalen Ansatz können wir auch Eingabeaufforderungen generieren, die ein Bild so beschreiben, dass wir es als Cartoon darstellen können. Durch die Eingabe eines Bilds zusammen mit einer Textbeschreibung kann das Modell seine visuellen Komponenten analysieren und dann eine kreative, detaillierte Eingabeaufforderung generieren, die speziell darauf ausgelegt ist, eine cartoonartige Darstellung zu erzeugen. Sobald wir diese Eingabeaufforderung haben, können wir sie an DALLE, das Bildgenerierungsmodell von OpenAI, weitergeben, um die visuellen Daten in eine stilisierte Cartoon-Version umzuwandeln. Dieser Workflow ermöglicht automatisierte, kreative Transformationen von Bildern und kombiniert die Interpretationsleistung eines multimodalen Modells mit den generativen Fähigkeiten von DALLE, sodass Benutzer benutzerdefinierte visuelle Ausgaben aus realen Bildern erstellen können.

````python
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI
import base64
import httpx
from openai import OpenAI
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

def modify_image(image_url: str, prompt: str):
    # Initialisieren des GPT-Modells
    model = ChatOpenAI(model="gpt-4o-mini")

    # Bilddaten abrufen und in Base64 kodieren
    image_data = base64.b64encode(httpx.get(image_url).content).decode("utf-8")

    # Erstellen Sie eine Nachricht mit Text und Bild
    message = HumanMessage(
        content=[
            {"type": "text", "text": prompt},
            {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{image_data}"}},
        ],
    )

    # Erhalten Sie eine Antwort mit einer geänderten Eingabeaufforderung von GPT
    response = model.invoke([message])
    cartoon_prompt = response.content

    # Initialisieren Sie das DALL-E-Modell, um das Bild zu generieren
    client = OpenAI()

    # Generieren Sie das Bild basierend auf der von GPT generierten Cartoon-Eingabeaufforderung
    response = client.images.generate(
        model="dall-e-3",
        prompt=cartoon_prompt,
        size="1024x1024",
        quality="standard",
        n=1,
    )

    # Holen Sie sich die Bild-URL von DALL-E
    image_url = response.data[0].url

    # Holen Sie sich das generierte Bild
    response2 = requests.get(image_url)
    img = Image.open(BytesIO(response2.content))

    # Zeigen Sie das Bild mit matplotlib an
    plt.imshow(img)
    plt.axis('off')  # Die Achsen ausblenden
    plt.show()

    # Gibt das Bildobjekt zurück, falls weitere Vorgänge erforderlich sind.
    return img

# Beispiel für den Aufruf der Funktion
image_url = 'https://data.heatonresearch.com/images/wustl/app_genai/hickory_deck_1.png'
# Bild-URL = 'https://www.heatonresearch.com/images/jeff/about-jeff-heaton-2020.jpg'

# prompt = "Gibt eine Eingabeaufforderung aus, die dieses Bild im Stil von Dali malt."
prompt = "Output a prompt that would render this image as a cartoon."

img = modify_image(image_url, prompt)
img.save("result.jpg")  # Speichern Sie das Bild als JPEG-Datei
````

# 3.5: LLM schreibt ein Buch

In einem früheren Modul haben wir einen LLM-basierten Buchgenerator erstellt, der die Erstellung eines Buches von Anfang bis Ende automatisiert. Dieser Prozess umfasste die Erstellung einer Zusammenfassung, die Gliederung des Inhaltsverzeichnisses und das iterative Schreiben der Kapitel in ein strukturiertes Markdown-Dokument. Jetzt werden wir diese Arbeit erweitern, indem wir DALLE integrieren, um ein Coverbild für das Buch zu generieren und ein visuelles Element hinzuzufügen, das den textbasierten Inhalt ergänzt.

Wir beginnen mit dem Zugriff auf ein großes Sprachmodell mit einer Temperatur von 0,7. Wir verwenden eine höhere Temperatur, um die Kreativität zu fördern.

````python
from langchain.chains.summarize import load_summarize_chain
from langchain import OpenAI, PromptTemplate
from langchain_openai import ChatOpenAI
from IPython.display import display_markdown

MODEL = 'gpt-4o-mini'

llm = ChatOpenAI(
        model=MODEL,
        temperature=0.7,
        n=1
    )
````

Wir erstellen eine einfache Hilfsfunktion, um den LLM mit einer benutzerdefinierten Systemeingabeaufforderung abzufragen. Die Systemeingabeaufforderung teilt dem LLM mit, dass er bei der Erstellung eines Buches hilft.

````python
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.prompts.chat import (
    ChatPromptTemplate,
    HumanMessagePromptTemplate,
    SystemMessagePromptTemplate,
)
from langchain_openai import ChatOpenAI

def query_llm(prompt):
  messages = [
      SystemMessage(
          content="You are assistant helping to write a book."
      ),
      HumanMessage(
          content=prompt
      ),
  ]

  output = llm.invoke(messages)
  return output.content
````


In diesem Abschnitt erweitern wir die Funktionalität unseres LLM-basierten Buchgenerators um eine visuelle Komponente: ein Coverbild für das Buch. Während sich unsere bisherige Arbeit auf die Generierung eines vollständigen Markdown-Dokuments mit der Inhaltsangabe, dem Inhaltsverzeichnis und den Kapiteln des Buches konzentrierte, führt diese Erweiterung die Leistungsfähigkeit von DALLE ein, um ein benutzerdefiniertes Coverbild basierend auf dem Inhalt des Buches zu generieren.

Um dies zu erreichen, wurde die Funktion render_dalle3 hinzugefügt. Durch Nutzung der OpenAI-API generieren wir mit DALLE 3 ein Bild basierend auf einer vorgegebenen Eingabeaufforderung, die der Titel oder ein beschreibender Aspekt des Buches sein könnte. Dieses Bild wird in der erforderlichen Auflösung von 1024 x 1024 Pixeln erstellt und kann an alle vom Benutzer angegebenen benutzerdefinierten Abmessungen angepasst werden. Das generierte Coverbild wird als JPEG-Datei gespeichert und kann optional mit der Python-Bibliothek matplotlib angezeigt werden.

Diese Funktion bereichert das erstellte Buch nicht nur mit einem professionellen Cover, sondern demonstriert auch die nahtlose Integration der Text- und Bildgenerierung, wodurch der Bucherstellungsprozess noch umfassender und optisch ansprechender wird.

````python
import io
import requests
from PIL import Image
import matplotlib.pyplot as plt
from openai import OpenAI

def render_dalle3(prompt, filename, height, width):
    # Initialisieren Sie den OpenAI-Client
    client = OpenAI()

    # Generieren Sie das Bild mit DALL-E 3 in der erforderlichen Auflösung von 1024 x 1024
    response = client.images.generate(
        model="dall-e-3",
        prompt=prompt,
        size="1024x1024",  # DALL-E 3 erfordert 1024x1024
        quality="standard",
        n=1,
    )

    # Holen Sie sich die Bild-URL aus der Antwort
    image_url = response.data[0].url

    # Holen Sie sich den Bildinhalt von der URL
    response2 = requests.get(image_url)
    img = Image.open(io.BytesIO(response2.content))

    # Passen Sie die Größe des Bildes an die gewünschten Abmessungen an (Höhe x Breite).
    img_resized = img.resize((width, height), Image.LANCZOS)

    # Speichern Sie das skalierte Bild
    img_resized.save(filename, "JPEG")

    # Optional das skalierte Bild anzeigen
    plt.imshow(img_resized)
    plt.axis('off')  # Blenden Sie die Achsen aus, um eine übersichtlichere Ansicht zu erhalten
    plt.show()

    return img_resized
````

## Erstellen Sie die Kapitel des Buches

````python
def render_chapter(num, chapter_title, title, subject, synopsis, toc):
  txt = query_llm(f"""
  Write Chapter {num}, titled "{chapter_title}" for a book of the title '{title}' for a book on
  the subject '{subject}' the book synopsis is '{synopsis}' the table of contents is '{toc}'.
  Give me only the chapter text, no chapter heading, no chapter title, number, no additional text.
  """).strip(" '\"")
  return txt

txt = render_chapter(1, toc2[0], title, SUBJECT, synopsis, toc)
print(txt)
````
