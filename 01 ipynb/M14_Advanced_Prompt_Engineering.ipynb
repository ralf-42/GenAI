# Teil 12.1: Einführung in Prompt Engineering

In der dynamischen Welt der Lebensversicherungen greifen Fachleute zunehmend auf künstliche Intelligenz zurück, um die Kundeninteraktion zu verbessern, Prozesse zu rationalisieren und schnell auf Informationen zuzugreifen. Prompt Engineering ist die Kunst, effektive Eingaben (Prompts) zu erstellen, um Sprachmodelle wie GPT-4 bei der Generierung nützlicher und genauer Ergebnisse zu leiten. Die Beherrschung von Prompt Engineering kann Lebensversicherungsfachleute in die Lage versetzen, präzise Informationen zu erhalten, überzeugende Kundenkommunikation zu erstellen und fundierte Entscheidungen zu treffen.

In diesem Leitfaden werden die wichtigsten Aspekte des Prompt Engineering für die Lebensversicherungsbranche behandelt. Er enthält praktische Beispiele, die Ihnen dabei helfen, KI-Sprachmodelle optimal zu nutzen.

### Geben Sie spezifische und vollständige Anweisungen

Durch spezifische und vollständige Anweisungen versteht das KI-Modell genau, was Sie benötigen. So werden Mehrdeutigkeiten reduziert und die Relevanz der Antwort erhöht.

Beispiel für eine ineffektive Eingabeaufforderung:

```Explain the types of life insurance policies```

Effective Prompt Example:

„Fassen Sie als Lebensversicherungsfachmann die wichtigsten Unterschiede zwischen der Kapitallebensversicherung und der Risikolebensversicherung zusammen und konzentrieren Sie sich dabei auf Leistungen, Laufzeit und typische Kundenprofile.“

Die effektive Eingabeaufforderung gibt die Rolle, die genauen benötigten Informationen und die Aspekte, auf die der Schwerpunkt gelegt werden soll, klar an und stellt sicher, dass die KI eine detaillierte und relevante Antwort liefert.

### Erklären Sie, was in einer Ausnahmesituation zu tun ist

Bedeutung: Indem Sie potenzielle Ausnahmen vorhersehen und dem Modell Anweisungen zum Umgang mit ihnen geben, können Sie unvollständige oder irreführende Antworten vermeiden.

```List the top five life insurance providers in the U.S. based on customer satisfaction. If current data is unavailable, please provide the most recent statistics from reputable sources and mention the year of the data.```

This prompt tells the AI what to do if up-to-date information isn't available, ensuring you still receive valuable data with proper context.

### Explain Output Format

Defining the desired output format makes the information easier to use and can facilitate integration with other documents or presentations.

```Erstellen Sie eine Vergleichstabelle für Risikolebensversicherungen und Kapitallebensversicherungen, einschließlich der Spalten „Policemerkmale“, „Vorteile“ und „Ideal für“. Präsentieren Sie die Informationen im Markdown-Format.```

Durch Angabe des Formats (einer Markdown-Tabelle) und der genauen benötigten Spalten kann die KI eine strukturierte und professionell aussehende Vergleichstabelle generieren, die für Kundenpräsentationen oder die interne Verwendung bereit ist.

### Bitten Sie um Vorschläge zu Ihrer Eingabeaufforderung

Durch die Anforderung von Feedback können Sie Ihre Eingabeaufforderungen verfeinern, was zu besseren Ergebnissen führt und Ihre Fähigkeiten im Bereich der Eingabeaufforderungsentwicklung verbessert.

```I want to draft an email to a client explaining the benefits of adding a critical illness rider to their life insurance policy. My current prompt is: 'Write an email about adding a critical illness rider.' Do you have suggestions to improve this prompt for more detailed and persuasive content?```

Possible AI Suggestion:

```Um Ihre Aufforderung zu verbessern, können Sie Einzelheiten zur Situation des Kunden und zu den spezifischen Vorteilen angeben, die für ihn von Bedeutung sein könnten. Beispiel: ‚Verfassen Sie eine überzeugende E-Mail an einen 40-jährigen Kunden mit einer Krankheitsgeschichte in der Familie und erklären Sie, wie er durch das Hinzufügen einer Zusatzversicherung für schwere Krankheiten zu seiner bestehenden Lebensversicherung im Falle schwerwiegender gesundheitlicher Probleme finanzielle Sicherheit erlangen kann.‘```

Die Vorschläge der KI sorgen für mehr Spezifität, was zu einer maßgeschneiderten und wirkungsvolleren E-Mail führen kann.

### Bedenken hinsichtlich Voreingenommenheit

Prompt Engineering kann strategisch eingesetzt werden, um die Voreingenommenheit in KI-generierten Antworten zu verringern, indem Eingabeaufforderungen sorgfältig entworfen werden, die Fairness und Neutralität fördern. Indem die KI explizit angewiesen wird, Stereotypen zu vermeiden, unterschiedliche Perspektiven zu berücksichtigen und eine inklusive Sprache zu verwenden, können Benutzer unbeabsichtigte Voreingenommenheiten abmildern, die in den Trainingsdaten des Modells vorhanden sein können. Beispielsweise könnte ein Lebensversicherungsfachmann Eingabeaufforderungen erstellen, die um Versicherungsempfehlungen bitten, ohne auf demografische Faktoren wie Alter, Geschlecht oder ethnische Zugehörigkeit zu verweisen, es sei denn, diese sind relevant und angemessen. Indem die an die KI gegebenen Anweisungen beachtet und der Bedarf an unvoreingenommenen und gerechten Informationen klargestellt wird, wird Prompt Engineering zu einem wertvollen Instrument zur Erzielung fairer und ausgewogener Ergebnisse.

# Teil 12.2: Ein paar Schüsse und Gedankengänge

In diesem Modul untersuchen wir zwei leistungsstarke Prompting-Techniken, die die Genauigkeit und den Nutzen generativer KI in Anwendungen wie der Risikoprüfung von Lebensversicherungen erheblich verbessern können: Few-Shot-Prompting und Chain-of-Thinking-Prompting. Beide Methoden befassen sich mit unterschiedlichen Herausforderungen, mit denen Versicherungsprompter konfrontiert werden können, wie z. B. der Bewertung des Antragstellerrisikos, der Erläuterung von Versicherungsoptionen oder der Schätzung von Prämien, indem die Antworten der KI so angepasst werden, dass sie durchdachter, genauer und an Branchenstandards ausgerichtet sind.

Few-Shot-Prompting, eine leistungsstarke Technik, gibt dem Modell einige Beispiele, um seine Reaktionen in neuen Situationen zu steuern. Diese Methode ist besonders effizient, wenn das Ziel darin besteht, strukturierte Ergebnisse zu generieren, wie etwa die Kategorisierung von Antragstellern nach Risiko, die Zusammenfassung von Policen oder die Schätzung der Lebenserwartung. Durch die Bereitstellung von nur einer Handvoll Beispiele lernt das Modell, mit ähnlichen Aufgaben umzugehen. Dies ist eine effiziente Möglichkeit, die Genauigkeit zu verbessern, ohne das Modell mit Daten zu überlasten. Im Kontext der Lebensversicherung kann Few-Shot-Prompting den Versicherern wertvolle Zeit sparen, indem sichergestellt wird, dass das KI-Modell Antworten produziert, die gut auf die jeweilige Aufgabe abgestimmt sind und den Mustern aus den Beispielen folgen.
Andererseits bietet die Gedankenkettenanregung den Versicherern ein wertvolles Werkzeug, um Einblick in den Entscheidungsprozess der KI zu gewinnen. Diese Technik ermöglicht es dem Modell, ein Problem Schritt für Schritt durchzugehen, was Transparenz in seinen Entscheidungsprozess bringt. Anstatt einfach nur eine Ausgabe anzubieten, geht das Modell seine Argumentation strukturiert durch. Diese Technik ist besonders wertvoll für komplexe Beurteilungen, bei denen mehrere Faktoren berücksichtigt werden müssen, wie z. B. die Bewertung der Gesundheitsgeschichte, des Lebensstils und der genetischen Veranlagung eines Antragstellers bei der Risikobestimmung. Indem die KI ermutigt wird, jeden Schritt zu artikulieren, erhalten die Versicherer Einblicke, wie das Modell zu seiner Schlussfolgerung gelangt ist, wodurch der Entscheidungsprozess zuverlässiger und erklärbarer wird.
Zusammen bieten Few-Shot-Prompting und Chain-of-Thought-Prompting Underwritern robuste Tools, um KI bei der Entscheidungsfindung zu nutzen. Few-Shot-Prompting ermöglicht es dem Modell, Muster anhand einiger bereitgestellter Beispiele zu replizieren, was bei Aufgaben wie Klassifizierung und Zusammenfassung hilft. Im Gegensatz dazu stellt Chain-of-Thought-Prompting sicher, dass komplexe Überlegungen gründlich und transparent sind. Diese Techniken können sich gegenseitig ergänzen und die Genauigkeit und Interpretierbarkeit von KI-generierten Antworten im Bereich Lebensversicherung verbessern.

### Eingabeaufforderung für wenige Aufnahmen

Im Zusammenhang mit Lebensversicherungen kann „Few-Shot“-Prompting ein leistungsstarkes Tool sein, um die Leistung generativer KI-Modelle für bestimmte Aufgaben zu optimieren. Few-Shot-Prompting bezeichnet die Technik, bei der dem Modell eine Handvoll Beispiele (normalerweise 2–5) gegeben werden, die die gewünschte Ausgabestruktur demonstrieren, bevor es eine neue Eingabe verarbeitet. Dies hilft dem Modell dabei, auf der Grundlage der bereitgestellten Beispiele genauere und kontextbezogen angemessenere Antworten zu generieren. Für Lebensversicherungs-Underwriter kann Few-Shot-Prompting Aufgaben wie die Analyse von Antragstellerdaten, die Erläuterung von Versicherungsoptionen oder die Schätzung von Risikofaktoren rationalisieren.

Ein Versicherer möchte beispielsweise, dass das Modell Antragsteller auf Grundlage ihrer Krankengeschichte in verschiedene Risikokategorien einteilt. Eine Aufforderung in wenigen Versuchen könnte wie folgt aussehen:

Eingabeaufforderungsbeispiel 1:
```
Classify the following applicants based on their medical history:

Applicant 1: Age 45, history of smoking, mild hypertension.
Classification: Moderate Risk.

Applicant 2: Age 30, non-smoker, no significant medical history.
Classification: Low Risk.

Applicant 3: Age 55, history of diabetes and heart disease.
Classification: High Risk.

Applicant 4: Age 50, history of high cholesterol.
Classification:
```

Durch die Einbeziehung mehrerer Beispiele von Antragstellern mit unterschiedlichen Gesundheitsprofilen ist das Modell in der Lage, die Beziehung zwischen Gesundheitsdaten und Risikokategorien zu verstehen. Die KI kann diese Muster dann auf neue Fälle anwenden und Versicherer dabei unterstützen, Risikofaktoren effizient zu identifizieren.

In einem anderen Fall kann eine KI mithilfe einer kurzen Eingabeaufforderung Richtlinienerklärungen erstellen. Ein Versicherer könnte Beispiele für einfache und klare Richtlinienzusammenfassungen liefern:

Eingabeaufforderungsbeispiel 2

```
Summarize these life insurance policies in a single sentence:

Policy A: $500,000 coverage, 30-year term, premiums increase after 10 years.
Summary: A 30-year term policy with $500,000 coverage, with premiums that increase after the first 10 years.
Policy B: $250,000 coverage, whole life policy with guaranteed premiums.
Summary: A whole life policy providing $250,000 coverage with guaranteed stable premiums.
Policy C: $350,000 coverage, CUL life policy.
Summary:
```

Diese Beispiele helfen dem Modell zu verstehen, wie man Versicherungsdetails auf prägnante und verständliche Weise zusammenfasst, was für Versicherer nützlich ist, die ihren Kunden komplexe Produkte erklären.

Ein letzter Anwendungsfall könnte die Schätzung der Lebenserwartung auf der Grundlage von Gesundheitsfaktoren sein, was ein entscheidender Teil der Risikobewertung für Lebensversicherungspolicen ist. Eine Aufforderung zu einigen Beispielen könnte folgende sein:

Eingabeaufforderungsbeispiel 3

```
Estimate the life expectancy for the following individuals based on their health profiles:

Person A: Age 60, non-smoker, exercises regularly, no significant medical conditions.
Life Expectancy: 85 years.
Person B: Age 45, smoker, history of high cholesterol and hypertension.
Life Expectancy: 70 years.
Person C: Age 35, heavy smoker, history of cancer and diabetes
Life Expectancy:
```

In diesem Fall liefert die Few-Shot-Prompting-Methode dem Modell klare Benchmarks für verschiedene Gesundheitsszenarien und hilft den Versicherern, die Lebenserwartung konsistenter einzuschätzen. Jedes Beispiel in der Eingabeaufforderung konditioniert das Modell, genaue Muster in seinen Zukunftsvorhersagen zu reproduzieren.

Durch die Verwendung von Few-Shot-Prompting auf diese Weise können Versicherer KI nutzen, um ihre Entscheidungsprozesse zu verbessern und gleichzeitig sicherzustellen, dass die Ergebnisse den Branchenstandards entsprechen.

### Gedankenkette

Bei der Denkketten-Anregung wird das KI-Modell dazu angehalten, seine Argumentation Schritt für Schritt zu artikulieren, bevor es zu einer Schlussfolgerung gelangt. Diese Methode trägt dazu bei, die Genauigkeit der Antworten des Modells zu verbessern, indem sie seinen Denkprozess transparenter und logischer macht. Im Kontext der Lebensversicherung kann die Denkketten-Anregung besonders bei komplexen Aufgaben wie Risikobewertung, Policenempfehlungen oder Prämienberechnungen nützlich sein, bei denen die Argumentation hinter einer Entscheidung genauso wichtig ist wie die Schlussfolgerung selbst.

Wenn ein Versicherer das Modell auffordert, das Risiko eines potenziellen Kunden anhand mehrerer Faktoren wie Alter, Gesundheitsgeschichte, Lebensstil und familiärer Krankengeschichte zu bewerten, kann das Modell mithilfe einer Denkkette jede Überlegung systematisch aufschlüsseln. Dadurch wird sichergestellt, dass kein entscheidendes Detail übersehen wird und die endgültige Entscheidung gut begründet ist.

```
Prompt:
Assess the risk level for a 52-year-old male applicant who has a history of hypertension, is a non-smoker, exercises regularly, and has a family history of heart disease. Please explain your reasoning step by step.

Model Response:
Let’s break down the risk assessment step by step:

Age: The applicant is 52 years old, which places him in the middle-aged category. While age alone does not immediately indicate high risk, it is a factor that needs to be considered, as health risks tend to increase with age.

Health history: The applicant has a history of hypertension. This condition is a significant risk factor for heart disease, stroke, and other cardiovascular problems. However, the severity of the hypertension and how well it is managed (e.g., through medication or lifestyle changes) would further influence this risk.

Lifestyle: The applicant is a non-smoker, which significantly reduces the risk of heart disease, lung disease, and cancer. Additionally, the fact that he exercises regularly is a positive factor, as regular physical activity helps control blood pressure, manage weight, and improve overall cardiovascular health. This healthy lifestyle choice can partially mitigate the risks posed by his hypertension.

Family medical history: The applicant has a family history of heart disease. This is an important consideration because it suggests a genetic predisposition to cardiovascular issues. Even though the applicant’s own lifestyle choices are healthy, this hereditary factor could increase his long-term risk.

Final assessment: Taking all factors into account, the applicant is not in the highest risk category due to his non-smoking status and regular exercise. However, his history of hypertension and family predisposition to heart disease moderately increase his risk. Therefore, I would classify him as moderate risk, with the recommendation to further monitor his blood pressure management and cardiovascular health over time.
```

In diesem Beispiel ermöglicht die Denkkette der KI, ihren Denkprozess klar zu artikulieren und jeden Faktor einzeln zu berücksichtigen, bevor sie zur endgültigen Risikoklassifizierung gelangt. Dieser Ansatz trägt dazu bei, sicherzustellen, dass die Versicherer der Logik der KI folgen können, und sorgt für Transparenz und einen strukturierteren Entscheidungsprozess.


# Teil 12.3: Persona- und Rollenmuster

Muster für Large Language Models (LLMs) sind strukturierte Techniken oder Vorlagen, die beim Erstellen von Eingabeaufforderungen verwendet werden, um bestimmte Arten von Antworten von der KI zu erhalten. Diese Muster steuern die Ausgabe des Modells, indem sie bestimmte Szenarien, Rollen oder Anweisungen innerhalb der Eingabeaufforderung einrichten. Durch die Verwendung von Mustern wie Personas, Rollenspielen oder semantischen Filtern können Benutzer die Konversation effektiv steuern und die Antworten der KI relevanter, kohärenter und auf bestimmte Bedürfnisse oder Kontexte zugeschnitten machen. Dieser Ansatz verbessert die Interaktion mit LLMs und ermöglicht eine präzisere und aussagekräftigere Kommunikation.

In diesem Teil werden wir untersuchen, wie Sie Interaktionen mit generativer KI mithilfe von Personas, Rollen und maßgeschneiderten Gesprächen auf der Grundlage definierter Identitäten im Lebensversicherungskontext gestalten können. Wenn Sie diese Muster beherrschen, können Sie ansprechendere und effektivere Eingabeaufforderungen erstellen, was zu umfassenderen und aussagekräftigeren KI-Antworten im Zusammenhang mit Lebensversicherungen führt.

### Persona-Muster
Beim Persona-Muster wird der KI ein bestimmter Charakter oder eine bestimmte Identität zugewiesen, sodass sie aus dieser Perspektive antworten kann. Diese Technik hilft dabei, den Ton, den Stil und den Inhalt der Antworten der KI zu formen.

Beispiel-Eingabeaufforderung:

```
You are an experienced life insurance agent specializing in family policies. Explain to a new client the benefits of whole life insurance.
```

### Zielgruppen-Persona
Die Audience Persona konzentriert sich darauf, die Antwort der KI auf ein bestimmtes Publikum zuzuschneiden. Indem Sie angeben, für wen die Informationen bestimmt sind, können Sie sicherstellen, dass der Inhalt für diese Gruppe angemessen und ansprechend ist.

Beispiel-Eingabeaufforderung:

```
Explain the importance of life insurance to a recent college graduate who has just entered the workforce.
```

### Rollenspiel-Eingabeaufforderungen
Rollenspiel-Aufforderungen binden die KI in ein simuliertes Szenario ein, in dem sie eine bestimmte Rolle übernimmt und so dynamische und kontextreiche Interaktionen ermöglicht.

Beispiel-Eingabeaufforderung:

```
You are a client interested in purchasing a life insurance policy. Initiate a conversation with an insurance agent, expressing your concerns and asking about policy options.
```

### Umgedrehtes Interaktionsmuster
Das Flipped Interaction Pattern kehrt die typischen Rollen um und fordert die KI auf, das Gespräch zu leiten oder Ihnen Fragen zu stellen. Dies kann kritisches Denken und aktives Engagement anregen.

Beispiel-Eingabeaufforderung:

```
You are my financial advisor. Ask me questions to help determine the best life insurance plan for my needs.
```

### Spielmuster
Das Game Play Pattern verwandelt die Interaktion in ein Spiel und macht den Austausch interaktiver und unterhaltsamer.

Beispiel-Eingabeaufforderung:

```
Let's play a quiz game about life insurance terms. Ask me five multiple-choice questions to test my knowledge.
```

### Anleitungen
Anleitungen weisen die KI an, klare, schrittweise Anweisungen oder Erklärungen zu einem bestimmten Thema oder Vorgang bereitzustellen.

Beispiel-Eingabeaufforderung:

```
Provide a step-by-step guide on how to file a life insurance claim after the loss of a loved one.
```

### Muster kombinieren
Beim Kombinieren von Mustern werden zwei oder mehr Eingabestrategien zusammen verwendet, um komplexe und differenzierte Interaktionen zu erstellen.

Beispiel-Eingabeaufforderung:

```
Assume the persona of a retirement planning expert (Persona Pattern) and explain to a couple in their 50s (Audience Persona) how life insurance can be a part of their retirement strategy.
```

### Semantisches Filtermuster
Das semantische Filtermuster weist die KI an, bestimmte Inhalte je nach Bedeutung ein- oder auszuschließen und stellt so sicher, dass die Antworten bestimmte Kriterien erfüllen oder bestimmte Themen vermeiden.

Beispiel-Eingabeaufforderung:

```
Summarize the key features of term life insurance policies, focusing only on benefits for young families and excluding any discussion of investment components.
```

Indem Sie diese Muster im Lebensversicherungskontext nutzen, können Sie KI-Interaktionen effektiv so steuern, dass sie persönlicher, ansprechender und für Ihren beabsichtigten Zweck geeigneter sind. Üben Sie das Erstellen von Eingabeaufforderungen mit diesen Techniken, um Ihre Kompetenz im Umgang mit generativer KI im Bereich Lebensversicherung zu verbessern.


# Teil 12.4: Frage-, Verfeinerungs- und Verifizierungsmuster

In diesem Teil beschäftigen wir uns mit Techniken zur Verfeinerung von Fragen, Validierung von Informationen und Eingabeaufforderungen auf der Grundlage iterativer Befragung und Überprüfung im Kontext der Lebensversicherung. Durch die Beherrschung dieser Muster können Sie genauere, detailliertere und zuverlässigere Informationen aus der generativen KI erhalten und so Ihre Interaktionen und Ergebnisse verbessern.

### Fragenverfeinerungsmuster
Beim Question Refinement Pattern beginnt man mit einer allgemeinen Frage und verfeinert sie dann schrittweise, um spezifischere und detailliertere Informationen zu erhalten. Diese Technik hilft dabei, tiefere Erkenntnisse zu gewinnen und komplexe Themen zu klären.

Beispiel-Eingabeaufforderung:

```
What are the main benefits of life insurance? Now, focus specifically on how life insurance can serve as an investment vehicle for retirement planning.
```

### Kognitives Prüfmuster
Das Cognitive-Verifier-Muster fordert die KI auf, Informationen zu verifizieren oder zu überprüfen, um die Genauigkeit und Zuverlässigkeit der Antwort sicherzustellen. Dieses Muster ist nützlich, um von der KI bereitgestellte Fakten oder Daten zu validieren.

Beispiel-Eingabeaufforderung:

```
Explain how universal life insurance policies work. Verify the information by summarizing the key features and any associated risks.
```

### Muster für eine Faktencheckliste
Das Fact Check List Pattern weist die KI an, Informationen im Format einer Checkliste bereitzustellen und so dabei zu helfen, Details zu überprüfen oder sicherzustellen, dass alle Aspekte eines Themas abgedeckt sind.

Beispiel-Eingabeaufforderung:

```
List the steps required to convert a term life insurance policy into a whole life policy, and briefly explain each step.
```

### Fragen-Antwort-Aufforderungen
Frage-Antwort-Eingabeaufforderungen sind direkte Abfragen, die darauf abzielen, bestimmte Informationen von der KI zu erhalten. Diese Eingabeaufforderungen ermöglichen klare und präzise Antworten auf bestimmte Fragen.

Beispiel-Eingabeaufforderung:

```
What tax advantages are associated with life insurance payouts to beneficiaries?
```

### Verfeinerungsaufforderungen
Bei Verfeinerungsaufforderungen wird die KI aufgefordert, eine vorherige Antwort zu verbessern oder zu präzisieren, um so die Tiefe und Qualität der bereitgestellten Informationen zu verbessern.

Beispiel-Eingabeaufforderung:

```
You mentioned that life insurance can help with estate planning. Can you provide more details on how it minimizes estate taxes?
```

### Vergleichs-/Kontrastaufforderungen
Bei Vergleichs-/Kontrastaufforderungen wird die KI aufgefordert, Ähnlichkeiten und Unterschiede zwischen zwei oder mehr Elementen hervorzuheben, um so ein klareres Verständnis der einzelnen Elemente zu ermöglichen.

Beispiel-Eingabeaufforderung:

```
Compare and contrast variable life insurance and universal life insurance in terms of investment options and risk.
```

### Bedingte Eingabeaufforderungen
Bedingte Eingabeaufforderungen weisen die KI an, Informationen basierend auf bestimmten Bedingungen oder Szenarien bereitzustellen und so maßgeschneiderte und relevante Antworten zu ermöglichen.

Beispiel-Eingabeaufforderung:

```
If a client is a smoker over the age of 50, how does that affect their life insurance premium and available policy options?
```

### Nach Eingabemuster fragen
Das Muster „Nach Eingabe fragen“ ermutigt die KI, zusätzliche Informationen oder Klarstellungen anzufordern und simuliert so eine interaktivere und persönlichere Konversation.

Beispiel-Eingabeaufforderung:

```
You are a life insurance advisor. Before making a recommendation, ask me questions to assess my coverage needs and financial situation.
```

Indem Sie diese Muster im Lebensversicherungskontext anwenden, können Sie Ihre Interaktionen mit Generative AI verfeinern, um präzisere, genauere und relevantere Informationen zu erhalten. Durch das Üben dieser Techniken verbessern Sie Ihre Fähigkeit, effektive Eingabeaufforderungen zu erstellen, was zu einer besseren Entscheidungsfindung und Kommunikation im Bereich Lebensversicherung führt.


# Teil 12.5: Inhaltserstellung und strukturierte Eingabeaufforderungsmuster

In diesem Teil behandeln wir, wie Sie Inhalte generieren und Antworten mithilfe bestimmter Formate oder Anweisungen strukturieren, um die Ergebnisse an den Lebensversicherungskontext anzupassen. Wenn Sie diese Muster beherrschen, können Sie Generative AI anleiten, Inhalte zu erstellen, die genau Ihren Anforderungen entsprechen – sei es beim Verfassen von Policenzusammenfassungen, beim Erstellen von Kundenkommunikation oder beim Extrahieren bestimmter Daten.

### Vorlagenmuster
Das Vorlagenmuster beinhaltet die Bereitstellung einer vordefinierten Struktur oder eines vordefinierten Formats, dem die KI bei der Generierung ihrer Antwort folgen soll. Dies ist besonders nützlich, wenn Sie Ausgaben in einem bestimmten Layout oder Stil benötigen.

Beispiel-Eingabeaufforderung:

```
Using the following template, create a life insurance policy summary:

Policy Name:
Policy Type:
Coverage Amount:
Premium Details:
Key Benefits:
Exclusions:
Additional Riders:
```

### Meta-Sprachmuster
Das Meta Language Pattern verwendet bestimmte Platzhalter oder Anweisungen innerhalb der Eingabeaufforderung, um die Antwort der KI zu steuern und so eine dynamische Inhaltsgenerierung zu ermöglichen.

Beispiel-Eingabeaufforderung:

```
Draft a personalized email to [Client Name], informing them about the benefits of upgrading their current life insurance policy to include [New Feature/Rider].
```

### Rezeptmuster
Das Rezeptmuster bietet schrittweise Anweisungen oder Verfahren und leitet die KI an, Inhalte zu erstellen, die einer logischen Abfolge folgen.

Beispiel-Eingabeaufforderung:

```
Provide a step-by-step guide on how to file a life insurance claim, including all necessary documentation and timelines.
```

### Muster für alternative Ansätze
Das Muster „Alternative Ansätze“ fordert die KI auf, mehrere Lösungen oder Perspektiven zu einem bestimmten Thema zu präsentieren, um ein breiteres Verständnis zu ermöglichen.

Beispiel-Eingabeaufforderung:

```
Describe three different strategies for a young family to maximize their life insurance coverage while staying within a limited budget.
```

### Gliederungserweiterungsmuster
Beim Outline Expansion Pattern erhalten Sie von der KI eine Gliederung oder eine Reihe von Aufzählungspunkten und werden gebeten, diese zu einer detaillierten Erzählung auszubauen.

Beispiel-Eingabeaufforderung:

```
Expand the following points into a comprehensive article about term life insurance:

Definition of term life insurance
Affordability and simplicity
Ideal candidates for term life policies
Comparison with whole life insurance
How to choose the right term length
```

### Menüaktionsmuster
Das Muster „Menüaktionen“ weist die KI an, als Reaktion auf ein Szenario Optionen oder Aktionselemente bereitzustellen, ähnlich der Auswahl aus einem Menü.

Beispiel-Eingabeaufforderung:

```
List the available policy options for a 35-year-old non-smoker seeking life insurance coverage of $500,000, including term lengths and premium estimates.
```

### Muster zur Schwanzgenerierung
Das Tail Generation Pattern weist die KI an, einen teilweise geschriebenen Inhalt fertigzustellen und so Konsistenz und Kontinuität sicherzustellen.

Beispiel-Eingabeaufforderung:

```
Finish the following letter to a client regarding their policy renewal:

'Dear [Client Name],

We wanted to remind you that your life insurance policy is up for renewal on [Date]. Continuing your coverage is important because...'
```

### Vervollständigungsaufforderungen
Vervollständigungsaufforderungen stellen den Anfang eines Satzes oder Absatzes dar und fordern die KI auf, ihn zu beenden. Diese werden oft verwendet, um Kreativität anzuregen oder Ideen zu generieren.

Beispiel-Eingabeaufforderung:

```
Life insurance plays a vital role in financial planning by...
```

### Eingabeaufforderungen zur Inhaltsgenerierung
Eingabeaufforderungen zur Inhaltserstellung fordern die KI auf, originelle Inhalte zu einem bestimmten Thema zu erstellen, die für die Erstellung von Artikeln, Berichten oder Marketingmaterialien nützlich sind.

Beispiel-Eingabeaufforderung:

```
Write a blog post titled 'The Top 5 Myths About Life Insurance Debunked' aimed at educating potential clients.
```

### Klassifizierungsaufforderungen
Bei Klassifizierungsaufforderungen geht es darum, Informationen anhand bestimmter Kriterien in Kategorien oder Gruppen zu organisieren.

Beispiel-Eingabeaufforderung:

```
Classify the following clients into appropriate life insurance policy types:

Client A: 28 years old, newly married, first child on the way
Client B: 45 years old, high-income earner, interested in investment opportunities
Client C: 60 years old, planning for estate transfer to grandchildren
```

### Übersetzungsaufforderungen
Übersetzungsaufforderungen fordern die KI auf, Inhalte von einer Sprache in eine andere zu übersetzen oder an einen anderen Stil oder ein anderes Format anzupassen.

Beispiel-Eingabeaufforderung:

```
Translate the following life insurance FAQ into simplified Chinese for our international clients:

'What is the difference between term and whole life insurance?'
```

### Eingabeaufforderungen zur Datenextraktion
Die Eingabeaufforderungen zur Datenextraktion weisen die KI an, bestimmte Informationen aus einem größeren Text zu extrahieren.

Beispiel-Eingabeaufforderung:

```
From the following policy document, extract the policyholder's name, policy number, coverage amount, and beneficiary details.
```

### Stilbasierte Eingabeaufforderungen
Stilbasierte Eingabeaufforderungen leiten die KI an, Inhalte in einem bestimmten Ton, Stil oder Tonfall zu erstellen und stellen sicher, dass die Ausgabe den Markenrichtlinien oder den Vorlieben des Publikums entspricht.

Beispiel-Eingabeaufforderung:

```
Write an informative article about the benefits of life insurance for small business owners in a professional and persuasive tone.
```

### Interaktive/iterative Eingabeaufforderungen
Interaktive/iterative Eingabeaufforderungen beinhalten einen fortlaufenden Austausch mit der KI, bei dem vorherige Antworten verfeinert und darauf aufgebaut werden, um ein gewünschtes Ergebnis zu erreichen.

Beispiel-Eingabeaufforderung:

```
Based on the following client profile, recommend a suitable life insurance policy. After your recommendation, I will provide feedback for you to adjust the proposal accordingly.

Client Profile:

Age: 40
Occupation: Teacher
Family: Married with two children
Financial Goals: Affordable premiums, college savings for children
```

Durch die Anwendung dieser Muster im Lebensversicherungskontext können Sie KI-generierte Inhalte effektiv strukturieren, um bestimmte Anforderungen zu erfüllen, die Kommunikation mit Kunden zu verbessern und die allgemeine Qualität der Informationen zu steigern. Üben Sie den Einsatz dieser Techniken, um generative KI so anzuleiten, dass sie in Ihrem Bereich maßgeschneiderte und qualitativ hochwertige Ergebnisse liefert.
