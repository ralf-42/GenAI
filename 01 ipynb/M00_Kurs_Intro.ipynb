# 2 | KI-Frameworks
---

Es gibt verschiedene Frameworks und Tools, die die Entwicklung von KI-Anwendungen erleichtern.

Zu den Top Frameworks/Tools gehören:


**Grundlegende Frameworks:**

+ LangChain (≥ 0.3+): Framework für die Entwicklung von LLM-basierten Anwendungen in Python
+ LangFlow: Separates visuelles Tool mit Low-Code-Interface, basierend auf LangChain
+ LlamaIndex: Datenframework für die Verbindung von LLMs mit verschiedenen Datenquellen
+ Hugging Face Transformers: Bibliothek für den Zugriff auf vortrainierte Transformer-Modelle

**Lokale LLM-Ausführung:**

+ Ollama: Tool für die lokale Ausführung von LLMs wie Llama 2 oder Falcon
+ LM Studio: Desktop-Anwendung mit grafischer Benutzeroberfläche für lokalen LLM

**Agenten-Frameworks:**

+ AutoGen: Framework für die Entwicklung autonomer Agenten
+ SmolAgents: Leichtgewichtiges Framework für Mikroagenten



Alle genannten Frameworks/Tools unterstützen Entwickler bei der Umsetzung von KI-Projekten und bieten verschiedene Möglichkeiten zur Integration großer Sprachmodelle.


Die drei Frameworks/Tools **LangChain**, **Ollama** und **Hugging Face Transformers**, die in diesem Kurs eingesetzt werden, bieten eine solide Basis für die praktische Anwendung generativer KI. In  Kombination mit **OpenAI** und den auf **Hugging Face verfügbaren Modellen** stehen modernste Werkzeuge zur Verfügung.
